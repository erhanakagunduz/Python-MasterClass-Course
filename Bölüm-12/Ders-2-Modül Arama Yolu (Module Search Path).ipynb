### Python Modül Arama Yolu
**Özet:** Bu derste, bir modülü programa içe aktarırken Python'daki modül arama yolunun nasıl çalıştığını öğreneceksiniz.

### Python Modül Arama Yoluna Giriş
Bir modülü bir programa içe aktardığınızda:

```python
import module
```

Python, aşağıdaki kaynaklardan `module.py` dosyasını arar:

- Programın çalıştırıldığı geçerli klasör.
- Önceden ayarladıysanız, PYTHONPATH ortam değişkeninde belirtilen klasörler listesi.
- Python'u kurarken yapılandırdığınız, kurulum bağımlı klasörler listesi.
  
Python, sonuçta oluşan arama yolunu `sys.path` değişkeninde saklar. Bu değişken, `sys` modülünden gelir.

Aşağıdaki program, geçerli modül arama yolunu gösterir:

```python
import sys

for path in sys.path:
    print(path)
```

**Windows'ta örnek bir çıktı:**

```
C:\python\python313.zip
C:\python\DLLs
C:\python\Lib
C:\python
C:\python\Lib\site-packages
```

Python'un her zaman `module.py` dosyasını bulabilmesini sağlamak için şunları yapmanız gerekir:

- `module.py` dosyasını programın çalışacağı klasöre yerleştirin.
- `module.py` dosyasını içeren klasörü PYTHONPATH ortam değişkenine ekleyin. Ya da `module.py` dosyasını PYTHONPATH değişkeninde bulunan klasörlerden birine yerleştirebilirsiniz.
- `module.py` dosyasını kurulum bağımlı klasörlerden birine yerleştirin.

### Python Modül Arama Yolunu Çalışma Zamanında Değiştirme
Python, `sys.path` değişkenini değiştirerek modül arama yolunu çalışma zamanında değiştirmenize izin verir. Bu, modül dosyalarını istediğiniz herhangi bir klasöre yerleştirmenizi sağlar.

`sys.path` bir liste olduğundan, arama yoluna yeni bir yol ekleyebilirsiniz.

Aşağıdaki örnek, `d:\modules` klasörünü arama yoluna ekler ve bu klasörde saklanan `recruitment` modülünü kullanır:

```python
>>> import sys
>>> sys.path.append('d:\\modules\\')
>>> import recruitment
>>> recruitment.hire()
Hire a new employee...
```

### Özet
Bir modül içe aktardığınızda, Python `module.py` dosyasını `sys.path` değişkeninde belirtilen klasörlerden arar.
Python, `sys.path` değişkenindeki öğeleri değiştirerek, ekleyerek veya kaldırarak modül arama yolunu değiştirmenize olanak tanır.
