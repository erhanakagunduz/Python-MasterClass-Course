### 1. **Modüllerin Kullanımı**
```python
# math modülünü kullanarak karekök hesaplayan program
import math

def calculate_square_root(number):
    return math.sqrt(number)

number = float(input("Bir sayı girin: "))
print(f"{number} sayısının karekökü: {calculate_square_root(number)}")
```

### 2. **Modül Arama Yolu (Module Search Path)**
```python
# sys modülünü kullanarak mevcut modül arama yolunu yazdırma
import sys

def print_search_path():
    for path in sys.path:
        print(path)

print_search_path()
```

### 3. **Python name Değişkeni**
```python
# __name__ değişkenini kullanarak modül çalıştırma örneği
def main():
    print("Ana fonksiyon çalıştı.")

if __name__ == '__main__':
    main()
```

### 4. **Paketler (Packages)**
```python
# sales paketi ve modülleri kullanarak örnek
# sales/order.py
def create_sales_order():
    print("Satış siparişi oluşturuldu.")

# sales/delivery.py
def create_delivery():
    print("Teslimat oluşturuldu.")

# main.py
from sales.order import create_sales_order
from sales.delivery import create_delivery

create_sales_order()
create_delivery()
```

### 5. 📌 **Bölüm Sonu Uygulaması**: math modülünü kullanarak karekök hesaplayan bir program
```python
# math modülünü kullanarak karekök hesaplama
import math

number = float(input("Bir sayı girin: "))
square_root = math.sqrt(number)
print(f"{number} sayısının karekökü: {square_root}")
```

### 6. 📌 **Orta Ölçekli Uygulama**: datetime modülü ile şu anki tarihi ve saati gösteren bir uygulama
```python
# datetime modülünü kullanarak şu anki tarihi ve saati gösteren program
from datetime import datetime

current_datetime = datetime.now()
print("Şu anki tarih ve saat:", current_datetime)
```
