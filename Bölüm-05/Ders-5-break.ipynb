### Python **break** İfadesi  
**Özet**: Bu derste, Python'daki **break** ifadesini ve bir döngüyü erken sonlandırmak için nasıl kullanılacağını öğreneceksiniz.

### Python break İfadesine Giriş  
Bazı durumlarda, bir **for** döngüsünü veya **while** döngüsünü, koşullu testlerin sonuçlarına bakılmaksızın erken sonlandırmak isteyebilirsiniz. Bu gibi durumlarda, **break** ifadesini kullanabilirsiniz:  
`break`

**Break** ifadesi genellikle, bir koşul doğru olduğunda döngüyü sonlandırmak için **if** ifadesiyle birlikte kullanılır.

### Python **break** ile **for** Döngüsü Kullanımı  
Aşağıdaki örnekte, **break** ifadesinin bir **for** döngüsünde nasıl kullanılacağı gösterilmektedir:

```python
for index in range(n):
    # diğer kodlar burada
    if condition:
        break
```

Bu sözdiziminde, eğer koşul **True** olarak değerlendirilirse, **break** ifadesi döngüyü hemen sonlandırır. Geriye kalan yinelemeleri çalıştırmaz.

Aşağıdaki örnek, **for** döngüsünde **break** ifadesinin nasıl kullanılacağını göstermektedir:

```python
for index in range(0, 10):
    print(index)
    if index == 3:
        break
```

**Çıktı**:

```
0
1
2
3
```

**Çalışma Prensibi**:  
Bu **for** döngüsü, 0'dan 9'a kadar 10 sayıyı sırasıyla ekrana yazdırır. Ancak, döngü sayacı 3 olduğunda, **break** ifadesi döngüyü hemen sonlandırır. Bu nedenle program yalnızca 0'dan 3'e kadar olan sayıları ekrana gösterir.

Eğer **break** ifadesini iç içe geçmiş döngülerde kullanırsanız, yalnızca en içteki döngüyü sonlandırır. Örneğin:

```python
for x in range(5):
    for y in range(5):
        # en içteki döngüyü sonlandır
        if y > 1:
            break
        # ekranı koordinatları göster
        print(f"({x},{y})")
```

**Çıktı**:

```
(0,0)
(0,1)
(1,0)
(1,1)
(2,0)
(2,1)
(3,0)
(3,1)
(4,0)
(4,1)
```

Bu örnekte, iki **for** döngüsü kullanılarak, koordinatlar (0,0) ile (5,5) arasında ekrana yazdırılmaktadır. İç içe döngüdeki **break** ifadesi, **y** değeri 1'den büyük olduğunda en içteki döngüyü sonlandırır. Bu nedenle yalnızca **y** değeri sıfır ve bir olan koordinatlar ekrana gelir.

### Python **break** İfadesi ile **while** Döngüsü Kullanımı  
Aşağıdaki örnekte, **while** döngüsünde **break** ifadesinin nasıl kullanılacağı gösterilmektedir:

```python
while condition:
    # diğer kodlar
    if condition:
        break
```

Aşağıdaki örnek, **break** ifadesinin bir **while** döngüsünde nasıl kullanılacağını göstermektedir:

Program, favori renginizi girmenizi ister ve "quit" komutu girildiğinde durur:

```python
print('-- Yardım: çıkmak için quit yazın --')
while True:
    color = input('Favori renginizi girin: ')
    if color.lower() == 'quit':
        break
```

**Çıktı**:

```
-- Yardım: çıkmak için quit yazın --
Favori renginizi girin: kırmızı
Favori renginizi girin: yeşil
Favori renginizi girin: mavi
Favori renginizi girin: quit
```

**Çalışma Prensibi**:  
**while True** ifadesi, sonsuz bir döngü oluşturur. "quit" komutunu girene kadar döngü devam eder. **color.lower() == 'quit'** koşulu **True** olduğunda, **break** ifadesi döngüyü sonlandırır. **color.lower()** komutu, kullanıcının "Quit", "QUIT" veya "quit" yazıp yazmadığına bakmaksızın tüm girişleri küçük harfe dönüştürür.

### Özet  
**Break** ifadesini kullanarak bir **for** döngüsünü veya **while** döngüsünü erken sonlandırabilirsiniz.
