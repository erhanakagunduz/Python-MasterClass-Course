### Python `for` Döngüsü ve `range()` Fonksiyonu

Bu derste, **Python for döngüsü** ve `range()` fonksiyonu kullanarak, bir kod bloğunu belirli bir sayıda nasıl çalıştırabileceğinizi öğreneceksiniz.

#### Python `for` Döngüsüne Giriş

Programlamada, bir kod bloğunu birden çok kez çalıştırmak isteyebilirsiniz. Bunu yapmak için **for döngüsü** kullanılır.

Aşağıda, **for döngüsünün** sözdizimi gösterilmektedir:

```python
for index in range(n):
    statement
```

Bu sözdiziminde:
- **index**: Döngü sayacıdır. 
- **n**: Döngünün çalışacağı toplam tekrarı belirtir.

Döngü sayacının adı **index** olmak zorunda değildir. İstediğiniz herhangi bir adı kullanabilirsiniz.

`range()` fonksiyonu, Python'da yerleşik bir fonksiyondur. **print()** fonksiyonu gibi, her zaman programda kullanılabilir.

`range(n)` fonksiyonu, sıfırdan başlayarak n'e kadar olan bir sayı dizisi oluşturur. Bu dizinin her bir öğesi birer birer artar ve n'ye ulaşana kadar devam eder.

Yani, `range(n)` şu sayılar dizisini üretir: 0, 1, 2, …, n-1. Unutmayın ki, son sayı olan **n** dizide yer almaz.

Aşağıdaki diyagramda for döngüsü ifadesi gösterilmektedir

````python
from IPython.display import Image

# Aynı dizindeki bir resmi görüntüleme
Image(filename="Python-for-loop.png")
````

Aşağıdaki örnek, **for döngüsü** ile `range()` fonksiyonunu kullanarak, ekrana 0'dan 4'e kadar 5 sayıyı nasıl yazdıracağınızı gösterir:

```python
for index in range(5):
    print(index)
```

**Çıktı:**

```
0
1
2
3
4
```

Bu örnekte, **for döngüsü** `print(index)` ifadesini tam beş kez çalıştırır.

Eğer ekrana 1'den 5'e kadar 5 sayıyı yazdırmak istiyorsanız, aşağıdaki gibi bir kod kullanabilirsiniz:

```python
for index in range(5):
    print(index + 1)
```

**Çıktı:**

```
1
2
3
4
5
```

Bu örnekte, her döngü iterasyonunda `index` değerini bir artırıyoruz ve bunu ekrana yazdırıyoruz. Ancak, bunu daha iyi bir şekilde yapmanın bir yolu vardır.

#### Dizinin Başlangıç Değerini Belirtme

Varsayılan olarak, `range()` fonksiyonu dizinin başlangıç değeri olarak sıfırı kullanır.

Bunun yanı sıra, `range()` fonksiyonu başlangıç değerini şu şekilde belirlemenize de olanak tanır:

```python
range(start, stop)
```

Bu sözdiziminde, `range()` fonksiyonu başlangıç değerini birer birer artırarak `stop` değerine kadar gider.

Aşağıdaki örnek, **for döngüsü** kullanarak 1'den 5'e kadar olan sayıları ekrana yazdırır:

```python
for index in range(1, 6):
    print(index)
```

**Çıktı:**

```
1
2
3
4
5
```

#### Dizinin Artış Miktarını Belirtme

Varsayılan olarak, `range(start, stop)` fonksiyonu başlangıç değerini her döngü iterasyonunda bir artırır.

Başlangıç değerini farklı bir sayıya artırmak için `range()` fonksiyonunun şu formunu kullanırsınız:

```python
range(start, stop, step)
```

Bu formda, `range()` fonksiyonunun hangi değere kadar artacağını belirtebilirsiniz.

Aşağıdaki örnek, 0'dan 10'a kadar olan tüm tek sayıları gösterir:

```python
for index in range(0, 11, 2):
    print(index)
```

**Çıktı:**

```
0
2
4
6
8
10
```

#### Python `for` Döngüsü Kullanarak Bir Dizinin Toplamını Hesaplama

Aşağıdaki örnek, **for döngüsü** kullanarak 1'den 100'e kadar olan sayıların toplamını hesaplar:

```python
sum = 0
for num in range(101):
    sum += num

print(sum)
```

**Çıktı:**

```
5050
```

**Nasıl çalışır:**
- İlk olarak, toplam değeri sıfır olarak başlatılır.
- Ardından, 1'den 100'e kadar olan her sayıya, her iterasyonda toplam eklenir.
- Son olarak, toplam ekrana yazdırılır.

Matematikçiler için basit bir formül de kullanılabilir:

```python
n = 100
sum = n * (n+1) / 2
print(sum)
```

### Özet

- **For döngüsünü**, bir kod bloğunu belirli sayıda çalıştırmak için kullanın.
- **range(start, stop, step)** fonksiyonunu kullanarak döngünüzü özelleştirebilirsiniz.
