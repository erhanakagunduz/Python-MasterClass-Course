### Python **continue** İfadesi  
**Özet**: Bu derste, Python'daki **continue** ifadesini ve döngüleri nasıl kontrol edeceğinizi öğreneceksiniz.

### Python **continue** İfadesine Giriş  
**Continue** ifadesi, bir **for** döngüsünde veya **while** döngüsünde kullanılır. **Continue** ifadesi, geçerli yinelemeyi atlar ve bir sonraki yinelemeye başlar.

Genellikle, bir koşul doğru olduğunda geçerli yinelemeyi atlamak için **continue** ifadesini **if** ifadesiyle birlikte kullanırsınız.

Aşağıda, **for** döngüsünde **continue** ifadesinin nasıl kullanılacağını gösteren bir örnek bulunmaktadır:

```python
for index in range(n):
    if condition:
        continue
    # diğer kodlar burada
```

Aşağıda ise **while** döngüsünde **continue** ifadesinin nasıl kullanılacağına dair bir örnek bulunmaktadır:

```python
while condition1:
    if condition2:
        continue
    # diğer kodlar burada
```

### Python **continue** İfadesi ile **for** Döngüsü Örneği  
Aşağıdaki örnek, 0 ile 9 arasındaki çift sayıları ekrana yazdırmak için **for** döngüsünde **continue** ifadesinin nasıl kullanılacağını göstermektedir:

```python
for index in range(10):
    if index % 2:
        continue

    print(index)
```

**Çıktı**:

```
0
2
4
6
8
```

**Çalışma Prensibi**:  
İlk olarak, **range()** fonksiyonu ile 0 ile 9 arasındaki sayılar üzerinden döngü oluşturulur.  
İkinci olarak, **index % 2** ifadesi ile **index** tek bir sayıysa (yani 1 döndürürse) geçerli yineleme atlanır ve bir sonraki yinelemeye geçilir. Bu sayede yalnızca çift sayılar ekrana yazdırılır.

### Python **continue** İfadesi ile **while** Döngüsü Örneği  
Aşağıdaki örnek, 0 ile 9 arasındaki tek sayıları ekrana yazdırmak için **while** döngüsünde **continue** ifadesinin nasıl kullanılacağını göstermektedir:

```python
# Tek sayıları yazdır
counter = 0
while counter < 10:
    counter += 1

    if not counter % 2:
        continue

    print(counter)
```

**Çıktı**:

```
1
3
5
7
9
```

**Çalışma Prensibi**:  
İlk olarak, **counter** değişkeni sıfır değeri ile başlatılır.  
Döngü, **counter** değeri 10'dan küçük olduğu sürece çalışır.  
Her yinelemede, **counter** bir artırılır. Eğer **counter** çift bir sayıysa, geçerli yineleme atlanır. Aksi takdirde, **counter** değeri ekrana yazdırılır.

### Özet  
Python'daki **continue** ifadesini kullanarak, geçerli yinelemeyi atlayabilir ve bir sonraki yinelemeye geçebilirsiniz. Bu, döngülerinizi belirli koşullara göre kontrol etmenize olanak tanır.
