### **Python `if` İfadesi**  

#### **Özet**  
Bu öğreticide, Python'da `if` ifadesinin nasıl kullanıldığını ve belirli bir koşula bağlı olarak bir kod bloğunun nasıl çalıştırılacağını öğreneceksiniz.  

#### **Basit Python `if` İfadesi**  
Python'da `if` ifadesi, belirli bir koşul sağlandığında bir kod bloğunu çalıştırmak için kullanılır.  

**Sözdizimi:**  
```python
if koşul:
    if_bloğu
```

Öncelikle, `if` ifadesi koşulu değerlendirir:  
- Koşul **`True`** (doğru) olarak değerlendirilirse, `if` bloğundaki ifadeler yürütülür.  
- Koşul **`False`** (yanlış) olarak değerlendirilirse, `if` bloğu atlanır.  

> **Dikkat:** Koşuldan sonra gelen iki nokta üst üste (`:`) işareti zorunludur. Eğer unutulursa, **`SyntaxError`** (sözdizimi hatası) oluşur.

### **Python `if` İfadesi Örneği**  

Aşağıdaki Python kodu, kullanıcıdan yaşını girmesini ister ve yaş 18 veya daha büyükse ekrana bir mesaj yazdırır:  

```python
age = input('Enter your age:')
if int(age) >= 18:
    print("You're eligible to vote.")
```

Bu kod şu şekilde çalışır:  
- Kullanıcıdan bir yaş girmesi istenir.  
- `int(age) >= 18` ifadesi, girilen değeri tam sayıya çevirir ve 18 ile karşılaştırır.  
- Eğer koşul **`True`** ise, `"You're eligible to vote."` mesajı ekrana yazdırılır.  
- Eğer koşul **`False`** ise, hiçbir işlem yapılmaz.  

#### **Örnek Çıktı:**  
```
Enter your age:18  
You're eligible to vote.
```

### **Girintilemenin (İndentation) Önemi**  
Python'da girintileme (indentation), kod bloklarının doğru çalışmasını sağlar. Örneğin:  

```python
age = input('Enter your age:')
if int(age) >= 18:
    print("You're eligible to vote.")
    print("Let's go and vote.")
```

Bu kodda, **`if` bloğuna ait olan iki satır** girintili yazılmıştır. Eğer koşul sağlanırsa, her iki mesaj da ekrana yazdırılır.

Ancak, aşağıdaki gibi bir kodda girintileme hatası, farklı bir davranışa neden olur:  

```python
age = input('Enter your age:')
if int(age) >= 18:
    print("You're eligible to vote.")
print("Let's go and vote.")
```

Bu kodda **"Let's go and vote."** mesajı, `if` bloğunun dışında olduğu için **her durumda ekrana yazdırılacaktır**.  

#### **Örnek Çıktı:**  
```
Enter your age:11  
Let's go and vote.
```

Bu nedenle, Python'da girintilemenin doğru kullanılması gereklidir. **`if` ifadesine ait tüm satırlar girintili olmalıdır.**

### **Python `if...else` İfadesi**  
Bazen, bir koşul sağlandığında bir işlem yapmak, sağlanmadığında ise farklı bir işlem yapmak isteyebilirsiniz. Bunun için **`if...else`** yapısını kullanabilirsiniz.  

**Genel sözdizimi:**  
```python
if koşul:
    if_bloğu
else:
    else_bloğu
```

Bu yapı şu şekilde çalışır:  
- Eğer **koşul `True`** ise, `if` bloğu çalıştırılır.  
- Eğer **koşul `False`** ise, `else` bloğu çalıştırılır.  

Bu yapı, **her durumda bir blok çalıştırılmasını garanti eder.**

## **Python `if…elif…else` İfadesi**
Birden fazla koşulu kontrol ederek belirli bir eylemi gerçekleştirmek için **`if...elif...else`** ifadesi kullanılır. **`elif`** ifadesi, **"else if"** anlamına gelir.

### **Genel Söz Dizimi:**
```python
if if-condition:
    if-block
elif elif-condition1:
    elif-block1
elif elif-condition2:
    elif-block2
...
else:
    else-block
```

📌 **Çalışma Mantığı:**  
1. Python, **if** koşulunu kontrol eder. **Doğru (`True`)** ise ilgili kod bloğunu çalıştırır ve aşağıdaki koşulları kontrol etmez.  
2. Eğer **if** koşulu **Yanlış (`False`)** ise, sıradaki **`elif`** koşulu kontrol edilir.  
3. **İlk `True` değerini döndüren koşulun kod bloğu çalıştırılır** ve kalan koşullar kontrol edilmez.  
4. Hiçbir koşul sağlanmazsa **else** bloğu çalıştırılır. (Bu blok **isteğe bağlıdır**.)  

---

### **Örnek Kullanım:**
```python
age = int(input("Enter your age: "))

if age < 12:
    print("You're a child.")
elif age < 18:
    print("You're a teenager.")
elif age < 65:
    print("You're an adult.")
else:
    print("You're a senior.")
```

### **Örnek Çıktılar:**
```
Enter your age: 10
You're a child.
```
```
Enter your age: 16
You're a teenager.
```
```
Enter your age: 30
You're an adult.
```
```
Enter your age: 70
You're a senior.
```

📌 **Önemli Not:**  
- **`if...elif...else`** yapısında, **ilk `True` olan koşul çalıştırılır ve geri kalan koşullar atlanır.**  
- **`else` bloğu opsiyoneldir.** Eğer eklenmezse ve hiçbir koşul sağlanmazsa program hiçbir şey yapmaz.

Aşağıdaki örnek, yaşa bağlı olarak bilet fiyatını belirlemek için **if...elif...else** deyimini kullanır:

```python
age = input('Enter your age:')

# String ifadeyi tam sayıya dönüştürme
your_age = int(age)

# Bilet fiyatını belirleme
if your_age < 5:
    ticket_price = 5
elif your_age < 16:
    ticket_price = 10
else:
    ticket_price = 18

# Bilet fiyatını gösterme
print(f"You'll pay ${ticket_price} for the ticket")
```

Bu örnekte:

- Eğer girilen yaş **5’ten küçükse**, bilet fiyatı **$5** olur.
- Eğer girilen yaş **5 veya daha büyük ve 16’dan küçükse**, bilet fiyatı **$10** olur.
- Aksi takdirde, bilet fiyatı **$18** olur.

### **Özet**
- **`if`** deyimini, bir koşula bağlı olarak bir kod bloğunu çalıştırmak için kullanın.
- **`if...else`** deyimini, koşul sağlanmadığında alternatif bir kod bloğu çalıştırmak için kullanın.
- **`if...elif...else`** deyimini, birden fazla koşulu kontrol ederek, **True** döndüren koşula bağlı olarak ilgili kod bloğunu çalıştırmak için kullanın.
