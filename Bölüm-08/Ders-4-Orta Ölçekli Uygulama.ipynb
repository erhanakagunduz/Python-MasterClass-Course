### **Orta Ölçekli Uygulama: Basit Bir Telefon Rehberi Uygulaması**

Telefon rehberi uygulaması, isim ve numara bilgilerini saklayan bir sözlük ile yapılabilir. Bu uygulamada kullanıcıdan isim ve numara bilgisi alacak ve bir telefon rehberi oluşturacağız:

```python
# Orta Ölçekli Uygulama: Telefon Rehberi Uygulaması

phone_book = {}

while True:
    print("\nTelefon Rehberi Uygulaması")
    print("1. Yeni Kişi Ekle")
    print("2. Kişiyi Ara")
    print("3. Rehberi Görüntüle")
    print("4. Çıkış")
    choice = input("Seçiminizi yapın (1/2/3/4): ")
    
    if choice == '1':
        name = input("Kişinin adını girin: ")
        phone_number = input(f"{name} için telefon numarasını girin: ")
        phone_book[name] = phone_number
        print(f"{name} rehberinize eklendi.")
    
    elif choice == '2':
        name = input("Aradığınız kişinin adını girin: ")
        if name in phone_book:
            print(f"{name}'in telefon numarası: {phone_book[name]}")
        else:
            print(f"{name} rehberde bulunamadı.")
    
    elif choice == '3':
        print("\nTelefon Rehberi:")
        for name, number in phone_book.items():
            print(f"{name}: {number}")
    
    elif choice == '4':
        print("Çıkılıyor...")
        break
    else:
        print("Geçersiz seçenek, tekrar deneyin.")
```
