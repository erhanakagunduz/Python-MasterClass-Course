### 2. Bir Dizin Ä°Ã§indeki DosyalarÄ± Listeleme
Bir dizindeki dosyalarÄ± listelemek iÃ§in `os.walk()` fonksiyonunu kullanabiliriz. Bu fonksiyon, belirtilen dizini ve alt dizinlerini gezerek dosyalarÄ±n yollarÄ±nÄ± dÃ¶ndÃ¼rÃ¼r.

### ğŸ“Œ BÃ¶lÃ¼m Sonu UygulamasÄ±: Mevcut Dizinindeki TÃ¼m DosyalarÄ± Listeleyen Python ProgramÄ±

AÅŸaÄŸÄ±da, mevcut Ã§alÄ±ÅŸma dizinindeki tÃ¼m dosyalarÄ± listeleyen basit bir Python programÄ± yer alÄ±yor:

```python
import os

# Mevcut Ã§alÄ±ÅŸma dizinini al
current_directory = os.getcwd()

# Mevcut dizindeki dosyalarÄ± listele
files = []

for dirpath, dirnames, filenames in os.walk(current_directory):
    for filename in filenames:
        files.append(os.path.join(dirpath, filename))

# DosyalarÄ± yazdÄ±r
for file in files:
    print(file)
```

### AÃ§Ä±klama:
1. `os.getcwd()` ile mevcut Ã§alÄ±ÅŸma dizini alÄ±nÄ±r.
2. `os.walk()` ile bu dizindeki tÃ¼m dosyalar gezilir.
3. Her dosyanÄ±n tam yolu `os.path.join()` ile belirlenir ve listeye eklenir.
4. Son olarak, tÃ¼m dosyalarÄ±n yolu yazdÄ±rÄ±lÄ±r.

### ğŸ“Œ Orta Ã–lÃ§ekli Uygulama: KullanÄ±cÄ±nÄ±n BelirttiÄŸi Dizinindeki TÃ¼m Dosya TÃ¼rlerini Tespit Eden Python ProgramÄ±

Bu uygulama, kullanÄ±cÄ±dan bir dizin yolu alacak ve belirtilen dizindeki tÃ¼m dosya tÃ¼rlerini (uzantÄ±larÄ±) tespit edecektir.

```python
import os

def list_file_extensions(path):
    """Bir dizindeki tÃ¼m dosya uzantÄ±larÄ±nÄ± listele"""
    extensions = set()  # TekrarlarÄ± Ã¶nlemek iÃ§in set kullanÄ±yoruz
    for dirpath, dirnames, filenames in os.walk(path):
        for filename in filenames:
            # DosyanÄ±n uzantÄ±sÄ±nÄ± al
            ext = os.path.splitext(filename)[1]
            extensions.add(ext)
    
    return extensions

# KullanÄ±cÄ±dan dizin yolu al
directory = input("Bir dizin yolu girin: ")

# Dizin geÃ§erli mi kontrol et
if os.path.isdir(directory):
    extensions = list_file_extensions(directory)
    if extensions:
        print("Bu dizindeki dosya tÃ¼rleri (uzantÄ±lar):")
        for ext in extensions:
            print(ext)
    else:
        print("Dizin boÅŸ.")
else:
    print("GeÃ§ersiz dizin yolu!")
```

### AÃ§Ä±klama:
1. `list_file_extensions()` fonksiyonu, `os.walk()` kullanarak belirtilen dizindeki dosyalarÄ± gezer.
2. Her dosyanÄ±n uzantÄ±sÄ± `os.path.splitext()` ile alÄ±nÄ±r ve `set` veri yapÄ±sÄ±na eklenir (bu sayede aynÄ± uzantÄ± birden fazla kez eklenmez).
3. KullanÄ±cÄ±dan dizin yolu alÄ±nÄ±r ve geÃ§erliliÄŸi kontrol edilir.
4. Son olarak, bulunan dosya tÃ¼rleri (uzantÄ±larÄ±) ekrana yazdÄ±rÄ±lÄ±r.

Bu uygulama, kullanÄ±cÄ±nÄ±n belirttiÄŸi dizindeki tÃ¼m dosya tÃ¼rlerini kolayca bulmanÄ±zÄ± saÄŸlar.
