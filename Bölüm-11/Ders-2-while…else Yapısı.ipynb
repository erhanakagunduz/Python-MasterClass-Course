#### **Python `while else` İfadesine Giriş**  
Python'da `while` döngüsünün isteğe bağlı bir `else` bloğu olabilir:

```python
while condition:
    # çalıştırılacak kod bloğu
else:
    # else bloğundaki kod
```

Bu sözdiziminde, koşul her iterasyonun başında kontrol edilir. `while` döngüsündeki kod bloğu, koşul **True** olduğu sürece çalışacaktır.

Koşul **False** olduğunda ve döngü normal şekilde sona erdiğinde, `else` bloğu çalışacaktır. Ancak, döngü `break` veya `return` ifadesi ile erken sona erdirilirse, `else` bloğu hiç çalışmaz.

### **Python `while else`**

**Özet:** Bu derste, Python `while else` ifadesini ve nasıl etkili bir şekilde kullanacağınızı öğreneceksiniz.

#### **Python `while else` İfadesi Örneği**

Diğer bazı programlama dillerinden (JavaScript, Java, C# gibi) geliyorsanız, bir döngüde `else` bloğunun biraz garip olduğunu düşünebilirsiniz.

Ancak, `while...else` bloğu bazı durumlarda çok faydalı olabilir. Şimdi `while...else` ifadesini kullanmaya yönelik bir örneğe bakalım.

Örneğin, aşağıdaki gibi meyvelerden oluşan bir listeyi ele alalım; her meyve, ad ve miktar gibi anahtarları içeren bir sözlüktür:

```python
basket = [
    {'fruit': 'apple', 'qty': 20},
    {'fruit': 'banana', 'qty': 30},
    {'fruit': 'orange', 'qty': 10}
]
```

Bu program, kullanıcıdan bir meyve ismi girmesini ister. Girdiği isme göre, `basket` listesinden meyve aranır ve bulunursa miktarı gösterilir. Eğer meyve bulunmazsa, kullanıcıya meyve için miktar girme imkanı tanınır ve yeni meyve listeye eklenir.

Aşağıdaki program ilk denemedir:

```python
basket = [
    {'fruit': 'apple', 'qty': 20},
    {'fruit': 'banana', 'qty': 30},
    {'fruit': 'orange', 'qty': 10}
]

fruit = input('Enter a fruit:')

index = 0
found_it = False

while index < len(basket):
    item = basket[index]
    # check the fruit name
    if item['fruit'] == fruit:
        found_it = True
        print(f"The basket has {item['qty']} {item['fruit']}(s)")
        break

    index += 1

if not found_it:
    qty = int(input(f'Enter the qty for {fruit}:'))
    basket.append({'fruit': fruit, 'qty': qty})
    print(basket)
```

**Nasıl Çalışır?**

1. Kullanıcıdan girdi almak için `input()` fonksiyonu kullanılır.
2. `index` sıfır olarak başlatılır ve `found_it` bayrağı False olarak ayarlanır. `index` listeye indeksle erişmek için, `found_it` bayrağı ise meyve adı bulunursa True yapılacaktır.
3. Liste üzerinde iterasyon yapılır ve meyve adı, girilen isimle eşleşip eşleşmediği kontrol edilir. Eğer eşleşirse, `found_it` bayrağı True yapılır, meyvenin miktarı gösterilir ve döngü `break` komutuyla sonlandırılır.
4. Döngü tamamlandıktan sonra `found_it` bayrağı kontrol edilir ve meyve bulunamadıysa, yeni meyve listeye eklenir.

Programın çıktısı şu şekilde olacaktır:

- Eğer "apple" girilirse:
  ```
  Enter a fruit:apple
  The basket has 20 apple(s)
  ```

- Eğer "lemon" girilirse:
  ```
  Enter a fruit:lemon
  Enter the qty for lemon:15
  [{'fruit': 'apple', 'qty': 20}, {'fruit': 'banana', 'qty': 30}, {'fruit': 'orange', 'qty': 10}, {'fruit': 'lemon', 'qty': 15}]
  ```

Program beklenildiği gibi çalışır.

Ancak, `while...else` ifadesi kullanıldığında, program çok daha kısa olacaktır.

Yeni sürümde `while...else` ifadesini kullanarak programı şu şekilde yazabiliriz:

```python
basket = [
    {'fruit': 'apple', 'qty': 20},
    {'fruit': 'banana', 'qty': 30},
    {'fruit': 'orange', 'qty': 10}
]

fruit = input('Enter a fruit:')

index = 0

while index < len(basket):
    item = basket[index]
    # check the fruit name
    if item['fruit'] == fruit:
        print(f"The basket has {item['qty']} {item['fruit']}(s)")
        found_it = True
        break

    index += 1
else:
    qty = int(input(f'Enter the qty for {fruit}:'))
    basket.append({'fruit': fruit, 'qty': qty})
    print(basket)
```

**Nasıl Çalışır?**

- Bu programda, `else` bloğu, `found_it` bayrağını ve döngü sonrasındaki `if` koşulunu ortadan kaldırır.
- Eğer meyve bulunmazsa, `while` döngüsü normal şekilde sona erer ve `else` bloğu çalışarak yeni meyve listeye eklenir.
- Ancak, meyve bulunursa, döngü `break` ifadesiyle sonlanır ve `else` bloğu çalışmaz.

### **Özet**

`while...else` ifadesindeki `else` bloğu, `while` döngüsü normal şekilde sona ererse çalışır ve `break` veya `return` ifadesi ile erken sonlandırılmadığı takdirde çalışacaktır. `while` döngüsünde bayrak kullanmanız gerektiğinde, Python `while else` ifadesini deneyin.
