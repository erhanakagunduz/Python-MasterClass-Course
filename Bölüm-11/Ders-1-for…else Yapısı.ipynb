### **Python `for‚Ä¶else` ƒ∞fadesine Giri≈ü**  

Python'da `for` d√∂ng√ºs√º, diƒüer programlama dillerinde (√∂rneƒüin Java veya C#) pek bulunmayan **isteƒüe baƒülƒ± bir `else` bloƒüu** i√ßerebilir.  

A≈üaƒüƒ±da `for...else` ifadesinin s√∂zdizimi g√∂sterilmektedir:  


üîπ **Bu yapƒ±nƒ±n √ßalƒ±≈üma mantƒ±ƒüƒ±:**  
‚úÖ **D√∂ng√º, `break` ifadesiyle kesilmezse**, `else` bloƒüu √ßalƒ±≈üƒ±r.  
‚úÖ **Eƒüer `break` kullanƒ±lƒ±rsa**, `else` bloƒüu atlanƒ±r.

### **Python `for‚Ä¶else` √ñrneƒüi**  

Diyelim ki, her bir elemanƒ± bir s√∂zl√ºk (`dictionary`) olan bir ki≈üi listemiz var. Her s√∂zl√ºk bir ki≈üinin adƒ±nƒ± ve ya≈üƒ±nƒ± i√ßeriyor:  

```python

          {'name': 'Jane', 'age': 22},
          {'name': 'Peter', 'age': 30},
          {'name': 'Jenifer', 'age': 28}]
```

Bir ki≈üinin adƒ±nƒ± aramak istiyoruz.  
- **Eƒüer liste i√ßerisinde ki≈üi bulunursa**, ki≈üinin bilgilerini ekrana yazdƒ±racaƒüƒ±z.  
- **Eƒüer ki≈üi yoksa**, "Ki≈üi bulunamadƒ±" mesajƒ± g√∂stereceƒüiz.  

Bunu geleneksel y√∂ntemle ≈üu ≈üekilde yapabiliriz:  

```python

          {'name': 'Jane', 'age': 22},
          {'name': 'Peter', 'age': 30},
          {'name': 'Jenifer', 'age': 28}]

name = 'Maria'

found = False
for person in people:
    if person['name'] == name:
        found = True
        print(person)
        break

if not found:
    print(f'{name} not found!')
```

üìå **√áƒ±ktƒ±:**  
```
Maria not found!
```

#### **Nasƒ±l √áalƒ±≈üƒ±yor?**  
1Ô∏è‚É£ **Aranacak adƒ± (`name`) belirliyoruz**.  
2Ô∏è‚É£ **`found` adƒ±nda bir bayrak (`flag`) deƒüi≈ükeni olu≈üturuyoruz**. Eƒüer ki≈üi bulunursa `True` yapƒ±p bilgiyi yazdƒ±rƒ±yoruz.  
3Ô∏è‚É£ **D√∂ng√ºden √ßƒ±ktƒ±ktan sonra** `found` deƒüi≈ükenine bakƒ±yoruz, eƒüer `False` ise "bulunamadƒ±" mesajƒ±nƒ± g√∂steriyoruz.  

üìå **Ancak `for‚Ä¶else` kullanarak bu programƒ± daha kƒ±sa yazabiliriz!**

### **`for‚Ä¶else` Nasƒ±l √áalƒ±≈üƒ±yor?**  
‚úÖ **Ki≈üi listede bulunursa:** Bilgi ekrana yazdƒ±rƒ±lƒ±r ve `break` ile d√∂ng√º sonlandƒ±rƒ±lƒ±r. `else` bloƒüu √ßalƒ±≈ümaz.  
‚úÖ **Ki≈üi listede yoksa:** D√∂ng√º normal ≈üekilde biter ve `else` bloƒüu √ßalƒ±≈üarak "bulunamadƒ±" mesajƒ±nƒ± yazdƒ±rƒ±r.  

‚ö° **Bu y√∂ntemin avantajƒ±, fazladan `found` bayraƒüƒ± ve if ko≈üulu kullanmamamƒ±zdƒ±r!**
