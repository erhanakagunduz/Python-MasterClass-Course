
# Python'da String'ler

## 🎯 Bu derste neler öğreneceksiniz?
- Python'da string nedir?
- String oluşturma, birleştirme, erişim ve dilimleme
- F-string ve özel karakter kullanımı
- String’lerin değiştirilemezliği (immutability)
 

## 🔤 String Nedir?
Bir *string* karakterlerden oluşan bir dizidir.
 
```python
mesaj = 'Bu bir metindir'
mesaj = "Ayrıca bu da bir metindir"
```

## 🧠 Tırnak Kullanımı
 
```python
mesaj  =  "'Bu bir stringtir'"       # Tek tırnak varsa çift tırnakla yaz
mesaj  =  '"İyilik iyidiri"'         # Çift tırnak varsa tek tırnakla yaz
```
 
## 🔄 Özel Karakterler ve Ham (Raw) String
 
```python
mesaj = 'Bu birinci satır\nBu ikinci satır'    # \n yeni satır karakteri
mesaj = 'Ali\'de geldi Ayşe\'de gelecek'       # \\ kaçış karakteri


Ham string (r): Kaçış karakterlerini olduğu gibi gösterir
mesaj = r'C:\python\bin'
```
 
## 📄 Çok Satırlı String
 
```python
help_message = '''
Usage: mysql command
     -h hostname
     -d database name
     -u username
     -p password
 '''
 print(help_message)
```

## 🔗 F-String Kullanımı
 
```python
 name = 'Erhan'
 message = f'Hi {name}'
 print(message)   # Hi Erhan
```

***F-string:*** Python 3.6+ ile geldi, string içinde değişkenleri gömülü yazabilirsiniz.
 
## 🔀 String Birleştirme

```python
 # 1. Literal string’ler otomatik birleşir
 greeting = 'Good ' 'Morning!'
 
 # 2. + operatörü ile birleşim
 greeting = 'Good '
 time = 'Afternoon'
 greeting = greeting + time + '!'
 print(greeting)  # Good Afternoon!
```
 
## 🔍 String Elemanlarına Erişim
 
```python
 str = "Python String"
 print(str[0])   # P
 print(str[1])   # y
 print(str[-1])  # g
 print(str[-2])  # n
```

## 📏 Uzunluk Ölçme

```python
 str = "Python String"
 print(len(str))  # 13
```

## ✂️ String Dilimleme (Slicing)

```python
 str = "Python String"
 print(str[0:2])  # Py
```

- `string[start:end]` → `start` dahil, `end` hariç
- `start` yazılmazsa → 0 kabul edilir
- `end` yazılmazsa → string sonuna kadar


## 🚫 String'ler Değiştirilemez (Immutable)

```python
 str = "Python String"
 str[0] = 'J'   # HATA! String'ler değiştirilemez
```

### Yeni string oluşturmak:

```python
 new_str = 'J' + str[1:]
 print(new_str)  # Jython String
```
## ✅ Özet
 
- String = karakter dizisi
- Değiştirilemez (immutable)
- `'...'`, `"..."`, `'''...'''` → string tanımlamada kullanılır
- `\` ile kaçış karakterleri, `r'...'` ile ham string
- `+` veya yan yana yazarak birleştirme
- `f"..."` ile değişken yerleştirme (f-string)
- `len()` → uzunluk
- `str[n]`, `str[start:end]` → erişim ve dilimleme
