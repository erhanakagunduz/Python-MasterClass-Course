### Python Comments  
**Özet:** Bu derste, Python'da kodunuza nasıl yorum ekleyeceğinizi ve çeşitli yorum türlerini öğreneceksiniz: blok yorumlar, satır içi yorumlar ve belge dizeleri.

#### Python Yorumları  
Kod yazarken bazen yazdığınız kodu açıklamak istersiniz. Örneğin, bir kod parçasının neden çalıştığını açıklamak isteyebilirsiniz. Bunu yapmak için yorumlar kullanırsınız.

Yorumlar, genellikle formülleri, algoritmaları ve karmaşık iş mantığını açıklamak için kullanılır.

Python yorumlar, program çalıştırılırken yorumları göz ardı eder ve yalnızca kodu yorumlar.

Python üç tür yorum sağlar: blok yorum, satır içi yorum ve belge dizeleri.

#### Python Blok Yorumlar  
Bir blok yorumu, onu takip eden kodu açıklar. Tipik olarak, bir blok yorumu, kod bloğuyla aynı seviyede girintilenir.

Blok yorumu oluşturmak için, tek bir `#` işareti ile başlarsınız, ardından bir boşluk ve bir metin dizisi eklersiniz. Örnek:

```python
# fiyatı %5 artır
price = price * 1.05
```

#### Python Satır İçi Yorumlar  
Bir yorumu, bir ifadeyle aynı satırda yazdığınızda, bu bir satır içi yorum olur.

Satır içi yorum da blok yorumla aynı şekilde başlar: tek bir `#` işareti, ardından bir boşluk ve bir metin dizisi.

Örnek:

```python
salary = 120000
salary = salary * 1.02   # maaşı %2 artır
print(salary)
```

#### Python Belge Dizeleri  
Bir belge dizisi, kod bloğunun ilk satırlarında, örneğin bir fonksiyonda yer alan bir dizedir.

Bir normal yorumdan farklı olarak, belge dizisini çalışma zamanında `obj.__doc__` özelliğini kullanarak erişebilirsiniz, burada `obj` fonksiyonun adı olur.

Belge dizileri, otomatik olarak kod belgelerini oluşturmak için kullanılır.

Python'da iki tür belge dizisi vardır: tek satırlık ve çok satırlı belge dizeleri.

#### Python Çok Satırlı Yorumlar  
Python çok satırlı yorumları desteklemez. Ancak, çok satırlı belge dizelerini çok satırlı yorum olarak kullanabilirsiniz. Python'un yaratıcısı Guido van Rossum da bunu önermiştir.

Yorumların net, öz ve açıklayıcı olması iyi bir pratiktir. Nihai hedef, zaman ve enerji tasarrufu sağlamaktır, böylece kodu daha sonra başkalarıyla çalışacak kişiler için de anlaşılır hale gelir.

#### Özet  
- Yorumlar, kodunuzu belgelendirmek için gereklidir.
- Bir blok yorumu ve satır içi yorum, `#` işaretiyle başlar.
- Fonksiyonlar, modüller ve sınıflar için belge dizeleri kullanılır.
