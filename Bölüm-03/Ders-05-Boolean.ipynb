### Python Boolean  
**Özet:** Bu derste, Python'daki boolean veri türünü, yanlış (False) ve doğru (True) değerleri öğreneceksiniz.

#### Python Boolean Veri Türü  
Programlamada, genellikle bir koşulun doğru olup olmadığını kontrol etmek ve sonuca göre bazı işlemler yapmak istersiniz.

Python, doğru ve yanlış değerlerini temsil etmek için boolean veri türünü sağlar. Boolean değeri, teknik olarak `bool` olarak adlandırılır.

Boolean veri türünün iki değeri vardır: **True** ve **False**.

Not: Boolean değerleri **True** ve **False** büyük harfle başlar (T ve F).

Aşağıdaki örnekte iki boolean değişkeni tanımlanmıştır:

```python
is_active = True
is_admin = False
```

#### Karşılaştırmalar ve Boolean Değerleri  
İki sayıyı karşılaştırdığınızda, Python sonucu boolean bir değer olarak döndürür. Örnek:

```python
x = 20
y = 10

result = x > y
print(result)

result = x < y
print(result)
```

Çıktı:

```
True
False
```

Ayrıca, iki stringi karşılaştırmak da boolean bir değer döndürür:

```python
x = 'a'
y = 'b'

result = x > y
print(result)

result = x < y
print(result)
```

Çıktı:

```
False
True
```

#### `bool()` Fonksiyonu  
Bir değerin True ya da False olup olmadığını öğrenmek için `bool()` fonksiyonunu kullanabilirsiniz. Örnek:

```python
result = bool('Hi')
print(result)

result = bool(100)
print(result)

result = bool(0)
print(result)
```

Çıktı:

```
True
True
False
```

Bu çıktı, bazı değerlerin True, bazı değerlerin ise False olduğunu gösterir.

#### Falsy ve Truthy Değerler  
Bir değer True olarak değerlendirilirse, o değeri "truthy" olarak adlandırırız. Eğer bir değer False olarak değerlendirilirse, o değeri "falsy" olarak adlandırırız.

Python'da **False** değerler şunlardır:
- Sayı sıfır (0)
- Boş string `''`
- `False`
- `None`
- Boş liste `[]`
- Boş tuple `()`
- Boş sözlük `{}`

**True** değerler ise falsy olmayan diğer tüm değerlerdir.

**Not:** None, liste, tuple ve sözlük gibi konuları ilerleyen derslerde daha detaylı inceleyeceksiniz.

#### Özet  
- Python boolean veri türü, iki değere sahiptir: **True** ve **False**.
- Bir değerin True ya da False olup olmadığını test etmek için `bool()` fonksiyonunu kullanabilirsiniz.
- **False** değerler False olarak değerlendirilirken, **True** değerler True olarak değerlendirilir.
- False değerler: sıfır, boş string, `False`, `None`, boş liste, boş tuple, ve boş sözlük.
- True değerler: False olmayan tüm diğer değerlerdir.
