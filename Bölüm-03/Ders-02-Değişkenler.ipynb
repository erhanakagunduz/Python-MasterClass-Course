### Python Değişkenleri  
**Özet:** Bu derste, Python değişkenlerini ve nasıl etkili bir şekilde kullanılacağını öğreneceksiniz.

#### Python'da Değişken Nedir?  
Bir program geliştirirken birçok değeri yönetmeniz gerekir. Bu değerleri saklamak için değişkenler kullanırsınız.

Python'da bir değişken, bir değeri atayabileceğiniz bir etikettir. Bir değişken her zaman bir değerle ilişkilidir. Örneğin:

```python
message = 'Hello, World!'
print(message)

message = 'Good Bye!'
print(message)
```

**Çıktı:**

```
Hello, World!
Good Bye!
```

Bu örnekte, `message` bir değişkendir. İlk satırda, 'Hello, World!' dizesini saklar ve `print()` fonksiyonu, bu mesajı ekrana yazdırır.

Sonraki satırda ise, 'Good Bye!' dizesi `message` değişkenine atanır ve ekrana yazdırılır.

`message` değişkeni, farklı zamanlarda farklı değerleri tutabilir ve program boyunca değerini değiştirebilir.

#### Değişkenler Oluşturma  
Bir değişken tanımlamak için aşağıdaki sözdizimini kullanırsınız:

```python
variable_name = value
```

Burada, `=` atama operatörüdür. Bu sözdiziminde, bir değeri `variable_name` değişkenine atıyorsunuz.

Değer, bir sayı, bir dize vb. herhangi bir şey olabilir.

Aşağıdaki örnek, `counter` adlı bir değişkeni tanımlar ve ona 1 değerini atar:

```python
counter = 1
```

#### Değişken İsimlendirme  
Bir değişken ismi verirken, bazı kurallara uymanız gerekir. Bu kurallara uymadığınızda hata alırsınız.

Aşağıdaki kurallar, değişkenler için dikkate almanız gereken kurallardır:

- Değişken isimleri yalnızca harfler, sayılar ve alt çizgi (_) içerebilir. Sayı ile başlayamazlar, harf veya alt çizgi ile başlamalıdır.
- Değişken isimleri boşluk içeremez. Birden fazla kelimeyi ayırmak için alt çizgi kullanılır. Örneğin: `sorted_list`.
- Değişken isimleri Python anahtar kelimeleri, rezerve kelimeler ve yerleşik fonksiyonlarla aynı olamaz.

Aşağıdaki yönergeler, iyi değişken isimleri oluşturmanıza yardımcı olur:

- Değişken isimleri kısa ve açıklayıcı olmalıdır. Örneğin, `active_user` değişkeni, `au`'dan daha açıklayıcıdır.
- Birden fazla kelimeyi ayırmak için alt çizgi (_) kullanın.
- Küçük "l" harfi ve büyük "O" harfini kullanmaktan kaçının, çünkü bunlar sırasıyla 1 ve 0 rakamına benzer.

#### Özet  
- Değişken, ona bir değer atayabileceğiniz bir etikettir. Bir değişkenin değeri, program boyunca değişebilir.
- Bir değişken oluşturmak için `variable_name = value` sözdizimini kullanın.
- Değişken isimleri mümkün olduğunca kısa ve açıklayıcı olmalı ve Python değişken isimlendirme kurallarına uymalıdır.
