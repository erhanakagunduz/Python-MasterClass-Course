**Python Tuple'ları**  
**Özet:** Bu derste, Python tuple'larını ve bunları nasıl etkin bir şekilde kullanacağınızı öğreneceksiniz.

---

### **Python Tuple'larına Giriş**  
Bazen, program boyunca değiştirilemeyen bir öğeler listesi oluşturmak istersiniz. Tuple'lar, bunu yapmanıza olanak tanır.

Bir tuple, değiştirilemeyen bir listedir. Python, değiştirilemeyen bir değeri "immutable" (değiştirilemez) olarak tanımlar. Bu nedenle, bir tuple, tanım gereği değiştirilemeyen bir listedir.

---

### **Tuple Tanımlamak**  
Bir tuple, bir listeye benzer ancak köşeli parantezler `[]` yerine parantezler `()` kullanır.

Aşağıdaki örnek, `rgb` adında bir tuple tanımlar:

```python
rgb = ('red', 'green', 'blue')
```

Bir tuple tanımlandıktan sonra, bireysel öğelere indeksle erişebilirsiniz. Örneğin:

```python
rgb = ('red', 'green', 'blue')

print(rgb[0])
print(rgb[1])
print(rgb[2])
```

**Çıktı:**

```python
red
green
blue
```

---

### **Bir Tuple Değiştirilemezdir**  
Bir tuple değiştirilemez olduğundan, elemanlarını değiştiremezsiniz. Aşağıdaki örnek, `rgb` tuple'ının ilk öğesini `'yellow'` olarak değiştirmeyi dener:

```python
rgb = ('red', 'green', 'blue')
rgb[0] = 'yellow'
```

Ve şu hata mesajını alırsınız:

```python
TypeError: 'tuple' object does not support item assignment
```

---

### **Bir Elemandan Oluşan Bir Tuple Tanımlamak**  
Bir elemandan oluşan bir tuple tanımlamak için, ilk öğenin sonrasında bir virgül (`,`) koymanız gerekir. Örneğin:

```python
numbers = (3,)
print(type(numbers))
```

**Çıktı:**

```python
<class 'tuple'>
```

Eğer son virgülü koymazsanız, `numbers`'ın tipi `int` olacaktır ve değeri `3` olacaktır. Python, `3` sayısının bulunduğu bir tuple oluşturmaz:

```python
numbers = (3)
print(type(numbers))
```

**Çıktı:**

```python
<class 'int'>
```

---

### **Bir Tuple Atamak**  
Bir tuple'ı değiştiremeseniz de, bir tuple'ı referans alan bir değişkene yeni bir tuple atayabilirsiniz. Örneğin:

```python
colors = ('red', 'green', 'blue')
print(colors)

colors = ('Cyan', 'Magenta', 'Yellow', 'black')
print(colors)
```

**Çıktı:**

```python
('red', 'green', 'blue')
('Cyan', 'Magenta', 'Yellow', 'black')
```

---

### **Özet**  
- Tuple'lar değiştirilemez (immutable) listelerdir.
- Değiştirilemeyen bir liste tanımlamak istediğinizde tuple kullanın.
