**Python'da Bir Liste Üzerinde For Döngüsü ile Döngü Yapma**  
**Özet:** Bu eğitimde, Python'da for döngüsünü kullanarak bir liste üzerinde nasıl döngü yapacağınızı öğreneceksiniz.

---

### **Python For Döngüsü ile Liste Üzerinde Döngü Yapmak**  
Bir liste üzerinde döngü yapmak için, aşağıdaki gibi for döngüsü kullanılır:

```python
for item in list:
    # item'ı işleyin
```

Bu söz diziminde, for döngüsü her yinelemede listenin bir elemanını `item` değişkenine atar.

Döngü bloğunda, her liste elemanını bireysel olarak işleyebilirsiniz.

Örneğin, aşağıdaki liste şehirleri tanımlar ve bu listeyi döngü ile iterasyona sokar:

```python
cities = ['New York', 'Pekin', 'Kahire', 'Mumbai', 'Meksika']

for city in cities:
    print(city)
```

**Çıktı:**

```
New York
Pekin
Kahire
Mumbai
Meksika
```

Bu örnekte, for döngüsü, `cities` listesinin her elemanını sırasıyla `city` değişkenine atar ve her iterasyonda bu şehri yazdırır.

---

### **For Döngüsünü İndeks ile Kullanmak**  
Bazen döngüdeki elemanların indekslerine erişmek isteyebilirsiniz. Bu durumlarda, `enumerate()` fonksiyonunu kullanabilirsiniz.

`enumerate()` fonksiyonu, listenin mevcut indeksini ve elemanını içeren bir tuple döndürür.

Aşağıdaki örnek, `enumerate()` fonksiyonu ile birlikte for döngüsünü kullanarak listeyi iterasyona sokar:

```python
cities = ['New York', 'Pekin', 'Kahire', 'Mumbai', 'Meksika']

for item in enumerate(cities):
    print(item)
```

**Çıktı:**

```
(0, 'New York')
(1, 'Pekin')
(2, 'Kahire')
(3, 'Mumbai')
(4, 'Meksika')
```

İndekse erişmek için, tuple'ı for döngüsü ifadesi içinde şu şekilde açabilirsiniz:

```python
cities = ['New York', 'Pekin', 'Kahire', 'Mumbai', 'Meksika']

for index, city in enumerate(cities):
    print(f"{index}: {city}")
```

**Çıktı:**

```
0: New York
1: Pekin
2: Kahire
3: Mumbai
4: Meksika
```

`enumerate()` fonksiyonu, başlangıç indeksini belirtmenize de olanak tanır. Varsayılan olarak sıfırdan başlar.

Aşağıdaki örnek, `enumerate()` fonksiyonunu ve indeksin birden başlamasını kullanır:

```python
cities = ['New York', 'Pekin', 'Kahire', 'Mumbai', 'Meksika']

for index, city in enumerate(cities, 1):
    print(f"{index}: {city}")
```

**Çıktı:**

```
1: New York
2: Pekin
3: Kahire
4: Mumbai
5: Meksika
```

---

### **Özet**  
- Bir liste üzerinde döngü yapmak için for döngüsü kullanın.
- İndeks erişimi için for döngüsü ile birlikte `enumerate()` fonksiyonunu kullanın.
