**Python filter() Fonksiyonu ile Liste Elemanlarını Filtreleme**  
**Özet:** Bu eğitimde, Python `filter()` fonksiyonunu kullanarak liste elemanlarını nasıl filtreleyeceğinizi öğreneceksiniz.

---

### **Python filter() Fonksiyonuna Giriş**  
Bazen bir listenin elemanları üzerinde yineleme yapmanız ve belirli bir kritere göre bazı elemanları seçmeniz gerekir.

Örneğin, aşağıdaki gibi bir puanlar listesine sahipsiniz:

```python
scores = [70, 60, 80, 90, 50]
```

Bu puanlar listesinden, her bir eleman 70'e eşit veya büyükse, o elemanı almak için şu kodu kullanabilirsiniz:

```python
scores = [70, 60, 80, 90, 50]

filtered = []

for score in scores:
    if score >= 70:
        filtered.append(score)

print(filtered)
```

**Çıktı:**

```
[70, 80, 90]
```

Bu nasıl çalışır?

1. İlk olarak, `filtered` adında boş bir liste tanımlanır; bu liste, `scores` listesinden elemanlar içerecek.
2. Daha sonra, `scores` listesi üzerinde yineleme yapılır. Eğer eleman 70 veya daha büyükse, bu eleman `filtered` listesine eklenir.
3. Son olarak, `filtered` listesi ekrana yazdırılır.

Python, bu işlemi daha şık bir şekilde yapabilmeniz için `filter()` adlı yerleşik bir fonksiyon sağlar.

### **filter() Fonksiyonunun Temel Kullanımı**  
`filter()` fonksiyonu, bir liste (veya demet) üzerindeki elemanları iteratif olarak işler ve her bir eleman için verilen fonksiyonu uygular. Fonksiyon, her eleman için `True` döndürürse, o eleman filtrelenmiş sonuca eklenir. `filter()` fonksiyonu bir iterator döndürür.

`filter()` fonksiyonunun temel sözdizimi şu şekildedir:

```python
filter(fn, list)
```

Burada `fn`, her liste elemanı için çağrılacak fonksiyonun adıdır.

Önceki örneği `filter()` fonksiyonu ile şöyle yazabiliriz:

```python
scores = [70, 60, 80, 90, 50]
filtered = filter(lambda score: score >= 70, scores)

print(list(filtered))
```

**Çıktı:**

```
[70, 80, 90]
```

`filter()` fonksiyonu bir iterator döndürdüğü için, üzerine bir `for` döngüsüyle yineleme yapabilirsiniz. Veya iterator'ı bir listeye dönüştürmek için `list()` fonksiyonunu kullanabilirsiniz.

### **filter() Fonksiyonunu Tuple Listesi ile Kullanma Örneği**  
Diyelim ki, aşağıdaki gibi bir ülkeler listesine sahipsiniz:

```python
countries = [
    ['China', 1394015977],
    ['United States', 329877505],
    ['India', 1326093247],
    ['Indonesia', 267026366],
    ['Bangladesh', 162650853],
    ['Pakistan', 233500636],
    ['Nigeria', 214028302],
    ['Brazil', 21171597],
    ['Russia', 141722205],
    ['Mexico', 128649565]
]
```

Her bir eleman, ülkenin adı ve nüfusunu içeren bir tuple'dır.

Eğer nüfusu 300 milyonun üzerinde olan ülkeleri almak istiyorsanız, `filter()` fonksiyonunu şu şekilde kullanabilirsiniz:

```python
countries = [
    ['China', 1394015977],
    ['United States', 329877505],
    ['India', 1326093247],
    ['Indonesia', 267026366],
    ['Bangladesh', 162650853],
    ['Pakistan', 233500636],
    ['Nigeria', 214028302],
    ['Brazil', 21171597],
    ['Russia', 141722205],
    ['Mexico', 128649565]
]

populated = filter(lambda c: c[1] > 300000000, countries)

print(list(populated))
```

**Çıktı:**

```
[['China', 1394015977], ['India', 1326093247], ['United States', 329877505]]
```

---

### **Özet**  
- Python `filter()` fonksiyonunu, bir liste (veya demet) üzerinde belirli bir kritere göre elemanları filtrelemek için kullanabilirsiniz.
- `filter()` fonksiyonu, belirli bir koşulu sağlayan elemanları içeren bir iterator döndürür.
