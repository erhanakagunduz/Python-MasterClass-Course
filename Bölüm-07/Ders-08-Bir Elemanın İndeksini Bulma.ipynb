**Python'da Bir Elemanın Listede İndeksini Bulma**  
**Özet:** Bu eğitimde, bir elemanın listede nasıl indeksini bulacağınızı öğreneceksiniz.

---

### **Bir Elemanın Listede İndeksini Bulmak**  
Bir elemanın listede indeksini bulmak için `index()` fonksiyonunu kullanırsınız.

Aşağıdaki örnek, şehirlerden oluşan bir liste tanımlar ve `index()` metodunu kullanarak 'Mumbai' değerine sahip elemanın indeksini bulur:

```python
cities = ['New York', 'Pekin', 'Kahire', 'Mumbai', 'Meksika']

result = cities.index('Mumbai')
print(result)
```

**Çıktı:**

```
3
```

Bu, beklenen şekilde 3 döndürür.

---

### **Listede Olmayan Bir Eleman Aramak**  
Ancak, listede olmayan bir elemanı `index()` fonksiyonu ile aramaya çalışırsanız bir hata alırsınız.

Aşağıdaki örnek, `index()` fonksiyonunu kullanarak `Osaka` şehrini arar:

```python
cities = ['New York', 'Pekin', 'Kahire', 'Mumbai', 'Meksika']

result = cities.index('Osaka')
print(result)
```

**Hata:**

```
ValueError: 'Osaka' is not in list
```

---

### **Hata Durumunda Çözüm**  
Bu sorunu çözmek için `in` operatörünü kullanmanız gerekir.

`in` operatörü, bir değerin listede olup olmadığını kontrol eder. Eğer varsa `True`, yoksa `False` döndürür.

`index()` fonksiyonunu kullanmadan önce, listede aradığınız elemanın olup olmadığını `in` operatörü ile kontrol edebilirsiniz. Örneğin:

```python
cities = ['New York', 'Pekin', 'Kahire', 'Mumbai', 'Meksika']
city = 'Osaka'

if city in cities:
    result = cities.index(city)
    print(f"{city} şehrinin indeks değeri: {result}.")
else:
    print(f"{city} liste içinde yok.")
```

**Çıktı:**

```
Osaka liste içinde yok.
```

---

### **Özet**  
- Bir elemanın listede indeksini bulmak için `in` operatörü ile `index()` fonksiyonunu kullanın.
