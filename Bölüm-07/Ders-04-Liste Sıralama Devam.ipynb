**Python sorted() Fonksiyonu**  
**Özet:** Bu derste, Python `sorted()` fonksiyonunu kullanarak bir listeyi nasıl sıralayacağınızı öğreneceksiniz.

---

### **Python sorted() Fonksiyonuna Giriş**  
`sort()` metodu, bir listeyi yerinde sıralar. Yani, orijinal listenin öğelerinin sırasını değiştirir.

Bir listeyi sıralayıp yeni sıralanmış listeyi elde etmek için `sorted()` fonksiyonunu kullanırsınız:

```python
sorted(list)
```

`sorted()` fonksiyonu, orijinal listeyi değiştirmez. Bunun yerine sıralanmış yeni bir liste döndürür.

Varsayılan olarak, `sorted()` fonksiyonu liste öğelerini küçükten büyüğe sıralar.

Sıralama sırasını tersine çevirmek için `reverse=True` argümanını şu şekilde kullanabilirsiniz:

```python
sorted(list, reverse=True)
```

---

### **String Listelerini Sıralama**  
Aşağıdaki örnek, `sorted()` fonksiyonunu kullanarak stringlerden oluşan bir listeyi alfabetik sıraya göre sıralar:

```python
guests = ['James', 'Mary', 'John', 'Patricia', 'Robert', 'Jennifer']
sorted_guests = sorted(guests)

print(guests)
print(sorted_guests)
```

**Çıktı:**

```python
['James', 'Mary', 'John', 'Patricia', 'Robert', 'Jennifer']
['James', 'Jennifer', 'John', 'Mary', 'Patricia', 'Robert']
```

Çıktıdan, orijinal listenin değişmediği ve `sorted()` fonksiyonunun sıralanmış yeni bir liste döndürdüğü görülmektedir.

Aşağıdaki örnek, `sorted()` fonksiyonu ile `guests` listesindeki öğeleri ters alfabetik sıraya göre sıralar:

```python
guests = ['James', 'Mary', 'John', 'Patricia', 'Robert', 'Jennifer']
sorted_guests = sorted(guests, reverse=True)

print(sorted_guests)
```

**Çıktı:**

```python
['Robert', 'Patricia', 'Mary', 'John', 'Jennifer', 'James']
```

---

### **Sayılardan Oluşan Bir Listeyi Sıralama**  
Aşağıdaki örnek, `sorted()` fonksiyonunu kullanarak `scores` listesindeki sayıları küçükten büyüğe sıralar:

```python
scores = [5, 7, 4, 6, 9, 8]
sorted_scores = sorted(scores)

print(sorted_scores)
```

**Çıktı:**

```python
[4, 5, 6, 7, 8, 9]
```

Aşağıdaki örnek, `sorted()` fonksiyonunu `reverse=True` argümanı ile kullanarak sayıları büyükten küçüğe sıralar:

```python
scores = [5, 7, 4, 6, 9, 8]
sorted_scores = sorted(scores, reverse=True)

print(sorted_scores)
```

**Çıktı:**

```python
[9, 8, 7, 6, 5, 4]
```

---

### **Özet**  
- `sorted()` fonksiyonunu, bir listeden yeni sıralanmış bir liste döndürmek için kullanın.
- `reverse=True` argümanı ile `sorted()` fonksiyonunu kullanarak bir listeyi ters sıralama yapabilirsiniz.
