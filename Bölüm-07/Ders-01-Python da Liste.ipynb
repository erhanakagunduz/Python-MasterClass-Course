Python, listeyi belirtmek için köşeli parantezleri (`[]`) kullanır. Aşağıda boş bir liste gösterilmektedir:


```python
todo_list = ['Python List Öğren', 'Liste elemanlarını nasıl yönetilir']
```

**Listedeki öğelere erişme**  
Bir liste sıralı bir koleksiyon olduğundan, elemanlarına indeksler aracılığıyla erişebilirsiniz:

```python
list[index]
```

**Liste sıfırdan başlayan indekslere sahiptir.** Diğer bir deyişle, ilk öğe `0` indeksine, ikinci öğe `1` indeksine ve devam eder.

Örneğin, aşağıdaki kodda `numbers` listesinin ilk elemanına nasıl erişileceği gösterilmektedir:

```python
numbers = [1, 3, 2, 7, 9, 4]
print(numbers[0])
```

**Çıktı:**

```python
1
```

`numbers[1]` komutu ise listedeki ikinci öğeyi döndürecektir:

```python
numbers = [1, 3, 2, 7, 9, 4]
print(numbers[1])
```

**Çıktı:**

```python
3
```

**Negatif indeksler**, listenin sonundan başlayarak öğelere erişmenizi sağlar.

- `list[-1]` son öğeyi döndürür.
- `list[-2]` sondan bir önceki öğeyi döndürür, ve devam eder. Örneğin:

```python
numbers = [1, 3, 2, 7, 9, 4]
print(numbers[-1])
print(numbers[-2])
```

**Çıktı:**

```python
4
9
```

**Öğeleri değiştirme, ekleme ve kaldırma**  
Bir liste dinamik bir yapıdır. Bu, liste elemanlarını değiştirebileceğiniz, yeni elemanlar ekleyebileceğiniz ve liste elemanlarını silebileceğiniz anlamına gelir.

### **Bir listedeki öğeleri değiştirme**  
Bir elemanı değiştirmek için, ona yeni bir değer atayarak bu işlemi gerçekleştirebilirsiniz:

```python
list[index] = new_value
```

Aşağıdaki örnek, `numbers` listesinin ilk elemanını 10 ile değiştirmeyi gösterir:

```python
numbers = [1, 3, 2, 7, 9, 4]
numbers[0] = 10

print(numbers)
```

**Çıktı:**

```python
[10, 3, 2, 7, 9, 4]
```

Aşağıdaki örnek, ikinci elemanı 10 ile çarpmayı gösterir:

```python
numbers = [1, 3, 2, 7, 9, 4]
numbers[1] = numbers[1] * 10

print(numbers)
```

**Çıktı:**

```python
[1, 30, 2, 7, 9, 4]
```

Ve aşağıdaki örnek, üçüncü elemanı 2'ye bölmeyi gösterir:

```python
numbers = [1, 3, 2, 7, 9, 4]
numbers[2] /= 2

print(numbers)
```

**Çıktı:**

```python
[1, 3, 1.0, 7, 9, 4]
```

### **Listeye eleman ekleme**  
`append()` metodu, bir öğeyi listenin sonuna ekler. Örneğin:

```python
numbers = [1, 3, 2, 7, 9, 4]
numbers.append(100)

print(numbers)
```

**Çıktı:**

```python
[1, 3, 2, 7, 9, 4, 100]
```

`insert()` metodu, bir öğeyi listenin herhangi bir konumuna ekler. Örneğin, aşağıdaki kodda, `100` sayısı `numbers` listesinin 2. indeksine eklenir:

```python
numbers = [1, 3, 2, 7, 9, 4]
numbers.insert(2, 100)

print(numbers)
```

**Çıktı:**

```python
[1, 3, 100, 2, 7, 9, 4]
```

---

### **Bir listeden öğeleri kaldırma**  
`del` komutu, bir öğeyi listeden silmenize olanak tanır, öğenin pozisyonunu belirterek bunu yaparsınız. Aşağıdaki örnekte, ilk öğe listeden silinir:

```python
numbers = [1, 3, 2, 7, 9, 4]
del numbers[0]

print(numbers)
```

**Çıktı:**

```python
[3, 2, 7, 9, 4]
```

`pop()` metodu, listenin son elemanını siler ve o elemanı döndürür:

```python
numbers = [1, 3, 2, 7, 9, 4]
last = numbers.pop()

print(last)
print(numbers)
```

**Çıktı:**

```python
4
[1, 3, 2, 7, 9]
```

**Genellikle**, `pop()` metodunu, bir öğeyi listeden çıkarmak ve aynı zamanda o öğenin değerine erişmek istediğinizde kullanırsınız.

Bir öğeyi indeksine göre silmek için, `pop()` metodunu indeks ile kullanırsınız. Örneğin:

```python
numbers = [1, 3, 2, 7, 9, 4]

second = numbers.pop(1)

print(second)
print(numbers)
```

**Çıktı:**

```python
3
[1, 2, 7, 9, 4]
```

Bir öğeyi değerine göre silmek için `remove()` metodunu kullanırsınız. `remove()` metodu, yalnızca listede karşılaştığı ilk öğeyi siler.

Örneğin, aşağıdaki kodda, `numbers` listesindeki `9` değeri silinir:

```python
numbers = [1, 3, 2, 7, 9, 4, 9]

numbers.remove(9)
print(numbers)
```

**Çıktı:**

```python
[1, 3, 2, 7, 4, 9]
```

Bu örnekte, `remove()` metodu yalnızca ilk `9` sayısını siler, ikinci `9` silinmez.

**Özet**  
- Liste, öğelerin sıralı bir koleksiyonudur.
- Köşeli parantez notasyonu (`[]`) kullanarak liste elemanına indeksle erişebilirsiniz. İlk öğe `0` indeksine sahiptir.
- Negatif indeks kullanarak, listenin sonundan başlayarak bir öğeye erişebilirsiniz. Son öğe `-1` indeksine sahiptir.
- `list[index] = new_value` ile bir öğeyi değiştirebilirsiniz.
- `append()` metodu, listeye yeni bir öğe ekler.
- `insert()` metodu, bir listeye belirtilen bir pozisyonda yeni bir öğe ekler.
- `pop()` metodu, bir öğeyi listeden çıkarır ve o öğeyi döndürür.
- `remove()` metodu, bir öğeyi listeden çıkarır.
