### Python Metin Dosyalarını Okuma

**Özet:** Bu eğitimde, Python'da metin dosyalarını okumanın çeşitli yollarını öğreneceksiniz.

**Kısa Özet:**  
Aşağıda, readme.txt dosyasındaki tüm metinlerin bir dize olarak nasıl okunacağını gösteren bir örnek yer almaktadır:

```python
with open('readme.txt') as f:
    lines = f.readlines()
```

### Python'da Metin Dosyasını Okuma Adımları
Python'da bir metin dosyasını okumak için şu adımları takip edersiniz:

1. **open() Fonksiyonu**  
   open() fonksiyonu birden fazla parametreye sahiptir ancak burada ilk iki parametreye odaklanacağız:

   ```python
   open(path_to_file, mode)
   ```

   - `path_to_file` parametresi, metin dosyasının yolunu belirtir.
   - Eğer program ile dosya aynı klasörde ise sadece dosya adını belirtmeniz yeterlidir. Aksi takdirde, dosyanın yolu ve dosya adı birlikte belirtilmelidir.
   - Dosya yolunu belirtirken, Windows kullanıyorsanız bile ileri eğik çizgi ('/') kullanmalısınız.

   Örneğin, readme.txt dosyası programla aynı klasörde ise şu şekilde açılır:

   ```python
   f = open('the-zen-of-python.txt', 'r')
   ```

   **Modlar (Modes):**
   - `'r'`: Dosyayı okuma modunda açar.
   - `'w'`: Dosyayı yazma modunda açar.
   - `'a'`: Dosyaya ekleme yapma modunda açar.

2. **Metin Okuma Yöntemleri**  
   Dosya nesnesi, metin dosyasından okuma işlemi yapmanızı sağlayacak üç yöntem sunar:

   - `read(size)`: Dosyanın içeriğini belirtilen boyutta okur ve bir dize olarak döndürür. Boyut belirtilmezse, dosyanın geri kalan kısmını okur.
   - `readline()`: Dosyadan tek bir satır okur ve bu satırı bir dize olarak döndürür.
   - `readlines()`: Dosyanın tüm satırlarını bir liste olarak okur. Küçük dosyalar için tüm metni manipüle etmek için kullanışlıdır.

3. **close() Yöntemi**  
   Dosyayı açtığınızda dosya, close() fonksiyonu ile kapatılana kadar açık kalır. Dosyayı kapatmak önemlidir çünkü:

   - Dosya açıldığında, genellikle dosya sistemi dosyayı kilitler, böylece başka programlar veya betikler dosyayı kullanamaz.
   - Dosya sistemi, oluşturabileceğiniz dosya tanıtıcılarının sayısına sınırlıdır.
   - Birçok dosyanın açık bırakılması, çoklu işlemlerin aynı dosyayı değiştirmeye çalıştığı yarış durumlarına yol açabilir ve beklenmedik davranışlar ortaya çıkabilir.

   Dosyayı kapatmak için şu kodu kullanabilirsiniz:

   ```python
   f.close()
   ```

   Dosyayı `close()` metodunu çağırmadan otomatik olarak kapatmak için `with` ifadesini şu şekilde kullanabilirsiniz:

   ```python
   with open(path_to_file) as f:
       contents = f.readlines()
   ```

   Gerçek hayatta, dosyayı otomatik olarak kapatmak için `with` ifadesi kullanılır.

### Metin Dosyasını Okuma Örnekleri

- **Tüm Dosyayı Okuma (read() Yöntemi):**

   ```python
   with open('the-zen-of-python.txt') as f:
       contents = f.read()
       print(contents)
   ```

   **Çıktı:**
   ```
   Beautiful is better than ugly.
   Explicit is better than implicit.
   Simple is better than complex.
   ...
   ```

- **Dosyayı Satır Satır Okuma (readlines() Yöntemi):**

   ```python
   with open('the-zen-of-python.txt') as f:
       [print(line) for line in f.readlines()]
   ```

   **Çıktı:**
   ```
   Beautiful is better than ugly.
   Explicit is better than implicit.
   Simple is better than complex.
   Complex is better than complicated.
   ...
   ```

- **Satırları Okurken Boş Satırları Kaldırma:**

   ```python
   with open('the-zen-of-python.txt') as f:
       [print(line.strip()) for line in f.readlines()]
   ```

- **Dosyayı Satır Satır Okuma (readline() Yöntemi):**

   ```python
   with open('the-zen-of-python.txt') as f:
       while True:
           line = f.readline()
           if not line:
               break
           print(line.strip())
   ```

   **Çıktı:**
   ```
   Explicit is better than implicit.
   Complex is better than complicated.
   Flat is better than nested.
   ...
   ```

- **Daha Kısa Bir Yöntemle Satır Satır Okuma:**

   ```python
   with open('the-zen-of-python.txt') as f:
       for line in f:
           print(line.strip())
   ```

### UTF-8 Metin Dosyalarını Okuma

Yukarıdaki örneklerde kullanılan yöntemler, ASCII metin dosyalarıyla gayet iyi çalışır. Ancak Japonca, Çince veya Korece gibi diğer dillerle çalışıyorsanız, dosya basit bir ASCII dosyası değildir ve muhtemelen UTF-8 formatında bir dosyadır.

UTF-8 metin dosyasını açmak için `encoding='utf-8'` parametresini `open()` fonksiyonuna eklemelisiniz:

```python
with open('quotes.txt', encoding='utf8') as f:
    for line in f:
        print(line.strip())
```

### Özet
- Bir metin dosyasını okumak için `open()` fonksiyonunu `'r'` modu ile kullanın.
- Metin dosyasını okumak için `read()`, `readline()`, veya `readlines()` metodlarını kullanın.
- Dosyayı okumayı tamamladıktan sonra her zaman `close()` metodunu veya `with` ifadesini kullanarak dosyayı kapatın.
- UTF-8 formatındaki bir metin dosyasını okumak için `encoding='utf-8'` parametresini kullanın.
