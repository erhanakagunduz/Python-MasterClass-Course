# **Python ile CSV Dosyası Yazma**  
**Özet:** Bu eğitimde, yerleşik `csv` modülünü kullanarak bir CSV dosyasına nasıl veri yazılacağını öğreneceksiniz.

### CSV Dosyası Yazma Adımları  
Bir CSV dosyasına veri yazmak için şu adımları takip etmelisiniz:

1. İlk olarak, `open()` fonksiyonunu kullanarak CSV dosyasını yazma (w modu) için açın.
2. İkinci olarak, `csv` modülünün `writer()` fonksiyonunu çağırarak bir CSV yazıcı nesnesi oluşturun.
3. Üçüncü olarak, CSV yazıcı nesnesinin `writerow()` veya `writerows()` metodunu çağırarak dosyaya veri yazın.
4. Son olarak, veri yazma işlemini tamamladıktan sonra dosyayı kapatın.

Aşağıdaki kod, yukarıdaki adımları göstermektedir:

```python
import csv

# Dosyayı yazma modunda aç
f = open('path/to/csv_file', 'w')

# CSV yazıcısı oluştur
writer = csv.writer(f)

# CSV dosyasına bir satır yaz
writer.writerow(row)

# Dosyayı kapat
f.close()
```

Dosyayı açıkça kapatmanıza gerek kalmaması için `with` ifadesini kullanarak daha kısa bir yazım yapabilirsiniz:

```python
import csv

# Dosyayı yazma modunda aç
with open('path/to/csv_file', 'w') as f:
    # CSV yazıcısı oluştur
    writer = csv.writer(f)

    # CSV dosyasına bir satır yaz
    writer.writerow(row)
```

Eğer ASCII olmayan karakterlerle çalışıyorsanız, `open()` fonksiyonunda karakter kodlamasını belirtmeniz gerekir.

Aşağıda, UTF-8 karakterlerinin bir CSV dosyasına nasıl yazılacağını gösteren bir örnek bulunmaktadır:

```python
import csv

# Dosyayı yazma modunda aç
with open('path/to/csv_file', 'w', encoding='UTF8') as f:
    # CSV yazıcısı oluştur
    writer = csv.writer(f)

    # CSV dosyasına bir satır yaz
    writer.writerow(row)
```

### CSV Dosyasına Yazma Örneği  
Aşağıdaki örnek, verileri bir CSV dosyasına nasıl yazacağınızı göstermektedir:

```python
import csv  

header = ['name', 'area', 'country_code2', 'country_code3']
data = ['Afghanistan', 652090, 'AF', 'AFG']

with open('countries.csv', 'w', encoding='UTF8') as f:
    writer = csv.writer(f)

    # Başlığı yaz
    writer.writerow(header)

    # Veriyi yaz
    writer.writerow(data)
```

`countries.csv` dosyasını açtığınızda, bir sorun göreceksiniz: Dosya içeriğinde iki ardışık satır arasında ek bir boş satır bulunmaktadır.

## **Birden fazla satır yazmak için CSV dosyasına**  
Birden fazla satırı aynı anda bir CSV dosyasına yazmak için, CSV yazıcı nesnesinin `writerows()` metodunu kullanırsınız.

Aşağıdaki örnek, `writerows()` metodunu kullanarak birden fazla satırı `countries.csv` dosyasına yazmaktadır:

```python
import csv

header = ['name', 'area', 'country_code2', 'country_code3']
data = [
    ['Albania', 28748, 'AL', 'ALB'],
    ['Algeria', 2381741, 'DZ', 'DZA'],
    ['American Samoa', 199, 'AS', 'ASM'],
    ['Andorra', 468, 'AD', 'AND'],
    ['Angola', 1246700, 'AO', 'AGO']
]

with open('countries.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)

    # Başlığı yaz
    writer.writerow(header)

    # Birden fazla satır yaz
    writer.writerows(data)
```

**CSV dosyasına DictWriter sınıfı ile yazmak**  
Eğer CSV dosyanızın her satırı bir sözlük (dictionary) ise, bu durumda `csv` modülünün `DictWriter` sınıfını kullanarak sözlükleri CSV dosyasına yazabilirsiniz.

Aşağıdaki örnek, `DictWriter` sınıfını kullanarak verileri bir CSV dosyasına yazmayı göstermektedir:

```python
import csv

# CSV başlığı
fieldnames = ['name', 'area', 'country_code2', 'country_code3']

# CSV verisi
rows = [
    {'name': 'Albania', 'area': 28748, 'country_code2': 'AL', 'country_code3': 'ALB'},
    {'name': 'Algeria', 'area': 2381741, 'country_code2': 'DZ', 'country_code3': 'DZA'},
    {'name': 'American Samoa', 'area': 199, 'country_code2': 'AS', 'country_code3': 'ASM'}
]

with open('countries.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()  # Başlığı yaz
    writer.writerows(rows)  # Satırları yaz
```

**Nasıl çalışır:**  
1. İlk olarak, CSV dosyasının alan isimlerini ve veri satırlarını tutan değişkenleri tanımlayın.  
2. Ardından, `open()` fonksiyonunu kullanarak CSV dosyasını yazma modunda açın.  
3. `DictWriter` sınıfının yeni bir örneğini oluşturun ve dosya nesnesini (`f`) ve `fieldnames` argümanını ona geçirin.  
4. Son olarak, `writeheader()` metodu ile başlığı yazın ve `writerows()` metodu ile veri satırlarını CSV dosyasına yazın.

**Özet**  
CSV dosyasına veri yazmak için `csv.writer()` veya `csv.DictWriter()` sınıfını kullanabilirsiniz.
