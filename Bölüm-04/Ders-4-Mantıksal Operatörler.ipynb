### Python Mantıksal Operatörleri  
**Özet:** Bu derste, Python mantıksal operatörlerini ve birden fazla koşulu nasıl birleştireceğinizi öğreneceksiniz.

#### Python Mantıksal Operatörlerine Giriş  
Bazen birden fazla koşulu aynı anda kontrol etmeniz gerekebilir. Bunu yapmak için mantıksal operatörleri kullanırsınız.

Python, üç mantıksal operatöre sahiptir:

- **and**  
- **or**  
- **not**

#### and Operatörü  
**and** operatörü, iki koşulun da aynı anda **True** olup olmadığını kontrol eder:

```python
a and b
```

Eğer her iki koşul da **True** ise **True** döndürür. Eğer bir koşul **False** ise **False** döndürür.

**Örnek:**

```python
price = 9.99
result = price > 9 and price < 10
print(result)  # True
```

**Çıktı:**
```
True
```

Eğer koşullardan biri **False** olursa, sonuç **False** olur:

```python
price = 9.99
result = price > 10 and price < 20
print(result)  # False
```

**Çıktı:**
```
False
```

**and** operatörünün sonucu aşağıdaki gibi bir tabloyla gösterilebilir:

| a     | b     | a and b |
|-------|-------|---------|
| True  | True  | True    |
| True  | False | False   |
| False | False | False   |
| False | True  | False   |

#### or Operatörü  
**or** operatörü de birden fazla koşulu kontrol eder. Ancak, herhangi bir koşul **True** olduğunda sonuç **True** döndürür. Eğer her iki koşul da **False** ise **False** döner:

```python
a or b
```

**Örnek:**

```python
price = 9.99
result = price > 10 or price < 20
print(result)  # True
```

**Çıktı:**
```
True
```

Her iki koşul da **False** olduğunda sonuç **False** olur:

```python
price = 9.99
result = price > 10 or price < 5
print(result)  # False
```

**Çıktı:**
```
False
```

**or** operatörünün sonucu aşağıdaki gibi bir tabloyla gösterilebilir:

| a     | b     | a or b |
|-------|-------|--------|
| True  | True  | True   |
| True  | False | True   |
| False | True  | True   |
| False | False | False  |

#### not Operatörü  
**not** operatörü, bir koşula uygulanır ve o koşulun sonucunu tersine çevirir. Eğer koşul **True** ise **False**, **False** ise **True** döndürür:

```python
not a
```

**Örnek:**

```python
price = 9.99
result = not price > 10
print(result)  # True
```

**Çıktı:**
```
True
```

**not** ve **and** operatörlerinin kombinasyonu:

```python
price = 9.99
result = not (price > 5 and price < 10)
print(result)  # False
```

**Çıktı:**
```
False
```

Burada Python, koşulları aşağıdaki sıraya göre değerlendirir:
1. **(price > 5 and price < 10)** ilk olarak **True** döndürür.
2. **not True** ise **False** döndürür.

#### Mantıksal Operatörlerin Önceliği  
Mantıksal operatörler bir ifade içinde karıştırıldığında, Python operatörleri belirli bir öncelik sırasına göre değerlendirir. İşte bu sıralama:

| Operatör | Öncelik |
|----------|---------|
| not      | Yüksek  |
| and      | Orta    |
| or       | Düşük   |

Bu öncelik sırasına göre Python, en yüksek öncelikli operatörle işlem yapmaya başlar. Aynı önceliğe sahip operatörler soldan sağa doğru değerlendirilir.

**Örnek:**

```python
a or b and c  # a or (b and c)
a and b or c and d  # (a and b) or (c and d)
a and b and c or d  # ((a and b) and c) or d
not a and b or c  # ((not a) and b) or c
```

#### Özet  
- Mantıksal operatörler, birden fazla koşulu birleştirmenizi sağlar.
- Python'da üç mantıksal operatör vardır: **and**, **or**, **not**.
- Mantıksal operatörlerin öncelik sırası: **not** > **and** > **or**.
