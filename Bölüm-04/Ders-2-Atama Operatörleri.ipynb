### Python Assignment Operators  
**Özet:** Bu derste, Python'daki atama operatörlerini kullanarak değişkenlere değer atamayı öğreneceksiniz.

#### Atama Operatörü (`=`)  
Python'da, bir değişkene değer atamak için atama operatörünü kullanırsınız. Örneğin, aşağıdaki kodda 0 sayısını `count` değişkenine atıyoruz:

```python
count = 0
print(count)
```

**Çıktı:**
```
0
```

Bir değişken üzerinde hesaplama yapıp sonucu aynı değişkene atayabilirsiniz:

```python
count = 0
count = count + 1
print(count)
```

**Çıktı:**
```
1
```

Burada, `count` değişkenine önce 0 değeri atandı, sonra değeri bir artırıldı ve tekrar `count`'a atandı.

Unutmayın, `=` işareti burada bir **atama** işlemi yapar, iki değeri karşılaştırmak için kullanılan eşitlik operatörü `==`'dir. 

#### Bileşik Atama Operatörleri  
Programlamada, genellikle fikirlerimizi daha kısa ve öz şekilde ifade etmek isteriz. Örneğin, `count` değişkenini artırıp sonucunu tekrar aynı değişkene atamak için bileşik atama operatörünü kullanabiliriz (`+=`):

```python
count = 0
count += 1  # Aynı şey: count = count + 1
print(count)
```

**Çıktı:**
```
1
```

Bu işlem, `count` değerini 1 artırır ve sonucu tekrar `count`'a atar. 

Python'daki yaygın bileşik atama operatörleri şunlardır:

| Operatör | Açıklama               | a  | b  | Örnek      | Eşdeğeri         | Sonuç |
|----------|------------------------|----|----|------------|------------------|-------|
| `+=`     | Topla ve Atama          | 5  | 2  | `a += b`   | `a = a + b`      | 7     |
| `-=`     | Çıkar ve Atama          | 5  | 2  | `a -= b`   | `a = a - b`      | 3     |
| `*=`     | Çarp ve Atama           | 5  | 2  | `a *= b`   | `a = a * b`      | 10    |
| `/=`     | Böl ve Atama            | 5  | 2  | `a /= b`   | `a = a / b`      | 2.5   |
| `//=`    | Tam Böl ve Atama        | 5  | 2  | `a //= b`  | `a = a // b`     | 2     |
| `%=`     | Modülüs ve Atama        | 5  | 2  | `a %= b`   | `a = a % b`      | 1     |
| `**=`    | Üssü Al ve Atama        | 5  | 2  | `a **= b`  | `a = a ** b`     | 25    |

#### Topla ve Atama (`+=`)  
Bu operatör, bir değeri değişkene ekler ve sonucu tekrar aynı değişkene atar. Örneğin:

```python
count = 0
count += 1  # Aynı şey: count = count + 1
print(count)
```

**Çıktı:**
```
1
```

#### Çıkar ve Atama (`-=`)  
Bu operatör, bir değeri değişkenden çıkarır ve sonucu tekrar aynı değişkene atar:

```python
quantity = 5
quantity -= 2  # Aynı şey: quantity = quantity - 2
print(quantity)
```

**Çıktı:**
```
3
```

#### Çarp ve Atama (`*=`)  
Bu operatör, bir sayıyı değişkenle çarpar ve sonucu tekrar aynı değişkene atar:

```python
a = 10
a *= 2  # Aynı şey: a = a * 2
print(a)
```

**Çıktı:**
```
20
```

#### Böl ve Atama (`/=`)  
Bu operatör, bir sayıyı değişkenle böler ve sonucu tekrar aynı değişkene atar:

```python
amount = 25
amount /= 2  # Aynı şey: amount = amount / 2
print(amount)
```

**Çıktı:**
```
12.5
```

#### Tam Böl ve Atama (`//=`)  
Bu operatör, tam bölme işlemi yapar ve sonucu aynı değişkene atar:

```python
amount = 10
amount //= 3  # Aynı şey: amount = amount // 3
print(amount)
```

**Çıktı:**
```
3
```

#### Modülüs ve Atama (`%=`)  
Bu operatör, modülüs (kalan) işlem yapar ve sonucu aynı değişkene atar:

```python
amount = 10
amount %= 3
print(amount)
```

**Çıktı:**
```
1
```

#### Üssü Al ve Atama (`**=`)  
Bu operatör, bir sayıyı üssüne yükseltir ve sonucu tekrar aynı değişkene atar:

```python
a = 2
a **= 3  # Aynı şey: a = a ** 3
print(a)
```

**Çıktı:**
```
8
```

#### Özet  
- Atama operatörü (`=`) bir değeri bir değişkene atamak için kullanılır.
- Bileşik atama operatörleri, bir işlem yapar ve sonucu aynı değişkene atar.
