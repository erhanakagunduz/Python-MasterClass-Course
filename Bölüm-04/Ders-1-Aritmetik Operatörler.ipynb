### Python Arithmetic Operators  
**Özet:** Bu derste, Python'daki aritmetik operatörleri kullanarak matematiksel işlemleri nasıl yapacağınızı öğreneceksiniz.

Python'da aritmetik operatörleri, sayılar üzerinde matematiksel işlemler yapmanıza olanak tanır. 
Aşağıdaki tabloda, Python'daki tüm aritmetik operatörler listelenmiştir:

| Operatör | Açıklama        | a  | b  | Örnek     | Sonuç  |
|----------|-----------------|----|----|-----------|--------|
| `+`      | Toplama         | 5  | 2  | `a + b`   | 7      |
| `-`      | Çıkarma         | 5  | 2  | `a - b`   | 3      |
| `*`      | Çarpma          | 5  | 2  | `a * b`   | 10     |
| `/`      | Bölme (Kesirli) | 5  | 2  | `a / b`   | 2.5    |
| `//`     | Tam Bölme       | 5  | 2  | `a // b`  | 2      |
| `%`      | Modülüs (Kalan) | 5  | 2  | `a % b`   | 1      |
| `**`     | Üssü Alma       | 5  | 2  | `a ** b`  | 25     |

#### Toplama (`+`)  
Toplama operatörü (`+`), iki sayıyı toplamak için kullanılır. Örneğin:

```python
a = 10
b = 3
result = a + b
print(result)
```

**Çıktı:**
```
13
```

#### Çıkarma (`-`)  
Çıkarma operatörü (`-`), ikinci sayıyı birinciden çıkararak farkı hesaplar. Örneğin:

```python
a = 10
b = 3
result = a - b
print(result)
```

**Çıktı:**
```
7
```

#### Çarpma (`*`)  
Çarpma operatörü (`*`), iki sayıyı çarpar. Örneğin:

```python
a = 10
b = 3
result = a * b
print(result)
```

**Çıktı:**
```
30
```

#### Bölme (`/`)  
Bölme operatörü (`/`), ilk sayıyı ikinci sayıya bölerek sonucu **kesirli** (float) olarak döndürür. Örneğin:

```python
a = 10
b = 3
result = a / b
print(result)
```

**Çıktı:**
```
3.3333333333333335
```

#### Tam Bölme (`//`)  
Tam bölme operatörü (`//`), bölme işlemi yapar fakat sonucu **tam sayı** olarak döndürür ve ondalıklı kısmı atar. Örneğin:

```python
a = 10
b = 3
result = a // b
print(result)
```

**Çıktı:**
```
3
```

#### Modülüs (`%`)  
Modülüs operatörü (`%`), iki sayının bölümünden kalanı döndürür. Örneğin:

```python
a = 10
b = 3
result = a % b
print(result)
```

**Çıktı:**
```
1
```

Modülüs operatörünü, bir sayının tek mi çift mi olduğunu kontrol etmek için de kullanabilirsiniz:

```python
a = 101

if a % 2 == 1:
    print(f'{a} is odd.')
else:
    print(f'{a} is even.')
```

**Çıktı:**
```
101 is odd.
```

Ayrıca, bir yılın artık yıl olup olmadığını kontrol etmek için de kullanabilirsiniz:

```python
year = 2024

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

**Çıktı:**
```
2024 is a leap year.
```

#### Üs Alma (`**`)  
Üs alma operatörü (`**`), ilk sayıyı ikinci sayıya üssünü alır. Örneğin:

```python
a = 10
b = 3
result = a ** b
print(result)
```

**Çıktı:**
```
1000
```

Üs alma operatörünü, finansal hesaplamalar ve bilimsel hesaplamalar için de kullanabilirsiniz. Örneğin, bileşik faiz hesaplaması yapabilirsiniz:

```python
principal = 1000
interest_rate = 0.05
year = 2
amount = principal * (1 + interest_rate) ** year

print(f"Total Amount: ${amount:,.2f}")
```

**Çıktı:**
```
Total Amount: $1,102.50
```

Bu örnekte, 1000 $'lık anapara, %5 faizle 2 yıl sonunda 1,102.50 $ olacaktır.

#### Özet  
- Python aritmetik operatörlerini kullanarak matematiksel işlemler yapabilirsiniz.
