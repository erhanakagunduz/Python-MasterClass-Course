### Python Karşılaştırma Operatörleri  
**Özet:** Bu derste, Python karşılaştırma operatörlerini ve iki değeri karşılaştırmak için nasıl kullanıldıklarını öğreneceksiniz.

#### Python Karşılaştırma Operatörlerine Giriş  
Programlamada, genellikle bir değeri başka bir değerle karşılaştırmak istersiniz. Bunu yapmak için karşılaştırma operatörlerini kullanırsınız.  

Python, altı karşılaştırma operatörüne sahiptir:

- **Küçük ( < )**
- **Küçük veya eşit ( <= )**
- **Büyük ( > )**
- **Büyük veya eşit ( >= )**
- **Eşit ( == )**
- **Eşit değil ( != )**

Bu karşılaştırma operatörleri, iki değeri karşılaştırır ve bir **boolean** (doğru veya yanlış) değeri döndürür. 

Bu operatörleri, hem sayılarla hem de dizelerle (string) kullanabilirsiniz.

#### Küçük ( < ) Operatörü  
Küçük (<) operatörü, sol değerin sağ değerden küçük olup olmadığını kontrol eder. Eğer küçükse **True**, aksi takdirde **False** döndürür:

```python
x = 10
y = 20

result = x < y
print(result)  # True

result = x > y
print(result)  # False
```

**Çıktı:**
```
True
False
```

Küçük (<) operatörünü dizelerle de kullanabilirsiniz. Karakterlerin ASCII sırasına göre karşılaştırma yapar:

```python
result = 'apple' < 'orange'
print(result)  # True

result = 'banana' < 'apple'
print(result)  # False
```

**Çıktı:**
```
True
False
```

#### Küçük veya Eşit ( <= ) Operatörü  
Küçük veya eşit (<=) operatörü, sol değerin sağ değere eşit veya küçük olup olmadığını kontrol eder:

```python
result = 20 <= 20
print(result)  # True

result = 10 <= 20
print(result)  # True

result = 30 <= 30
print(result)  # True
```

**Çıktı:**
```
True
True
True
```

#### Büyük ( > ) Operatörü  
Büyük (>) operatörü, sol değerin sağ değerden büyük olup olmadığını kontrol eder:

```python
result = 20 > 10
print(result)  # True

result = 20 > 20
print(result)  # False

result = 10 > 20
print(result)  # False
```

**Çıktı:**
```
True
False
False
```

#### Büyük veya Eşit ( >= ) Operatörü  
Büyük veya eşit (>=) operatörü, sol değerin sağ değere eşit veya büyük olup olmadığını kontrol eder:

```python
result = 20 >= 10
print(result)  # True

result = 20 >= 20
print(result)  # True

result = 10 >= 20
print(result)  # False
```

**Çıktı:**
```
True
True
False
```

#### Eşit ( == ) Operatörü  
Eşit (==) operatörü, iki değerin birbirine eşit olup olmadığını kontrol eder. Eğer eşitse **True**, değilse **False** döndürür:

```python
result = 20 == 10
print(result)  # False

result = 20 == 20
print(result)  # True
```

**Çıktı:**
```
False
True
```

#### Eşit Değil ( != ) Operatörü  
Eşit değil (!=) operatörü, iki değerin birbirine eşit olup olmadığını kontrol eder. Eğer eşit değilse **True**, eşitse **False** döndürür:

```python
result = 20 != 20
print(result)  # False

result = 20 != 10
print(result)  # True
```

**Çıktı:**
```
False
True
```

#### Özet  
- Karşılaştırma operatörleri, iki değeri karşılaştırarak **True** veya **False** döndürür.
- Python'da altı karşılaştırma operatörü vardır: `<`, `<=`, `>`, `>=`, `==`, `!=`.
