Python `issuperset()`  
Özet: Python'da bir setin başka bir setin süper kümesi olup olmadığını kontrol etmek için `issuperset()` metodunu kullanırsınız.

### `issuperset()` Metodu  
`set_a.issuperset(set_b)`  
`issuperset()` metodu, `set_a` seti `set_b` setinin süper kümesi ise `True`, aksi takdirde `False` döner.

### Python `issuperset()` Metodu Örnekleri  
Aşağıdaki örnek, `numbers` setinin `scores` setinin süper kümesi olup olmadığını kontrol eder:

```python
numbers = {1, 2, 3, 4, 5}
scores = {1, 2, 3}

result = numbers.issuperset(scores)

print(result)
```

**Çıktı:**

```
True
```

`numbers` seti, `scores` setindeki tüm elemanlara sahip olduğundan, `numbers` seti `scores` setinin süper kümesidir.

Bir set, kendisinin de süper kümesidir. Örneğin:

```python
numbers = {1, 2, 3, 4, 5}
result = numbers.issuperset(numbers)

print(result)
```

**Çıktı:**

```
True
```

Aşağıdaki örnekte, `scores` seti, `numbers` setinin alt kümesi olmadığı için sonuç `False` döner:

```python
numbers = {1, 2, 3, 4, 5}
scores = {1, 2, 3}

result = scores.issuperset(numbers)

print(result)
```

**Çıktı:**

```
False
```

### Süper Küme Operatörlerini Kullanma  
Bir setin başka bir setin süper kümesi olup olmadığını kontrol etmek için `>=` operatörünü kullanabilirsiniz:

```python
set_a >= set_b
```

`>=` operatörü, `set_a` seti `set_b` setinin süper kümesi ise `True`, aksi takdirde `False` döner. Örneğin:

```python
numbers = {1, 2, 3, 4, 5}
scores = {1, 2, 3}

result = numbers >= scores
print(result)  # True

result = numbers >= numbers
print(result)  # True
```

**Çıktı:**

```
True
True
```

Bir setin başka bir setin doğru süper kümesi olup olmadığını kontrol etmek için `>` operatörünü kullanırsınız:

```python
set_a > set_b
```

Örneğin:

```python
numbers = {1, 2, 3, 4, 5}
scores = {1, 2, 3}

result = numbers > scores
print(result)  # True

result = numbers > numbers
print(result)  # False
```

**Çıktı:**

```
True
False
```

Bu örnekte, `numbers` seti kendisinin doğru süper kümesi olmadığı için, `>` operatörü `False` döner.

### Özet  
- Set A, Set B'nin süper kümesi ise, Set B'nin tüm elemanları Set A'da bulunur.
- Bir setin başka bir setin süper kümesi olup olmadığını kontrol etmek için `issuperset()` metodunu kullanın.
- Süper küme operatörü (`>=`) veya doğru süper küme operatörü (`>`) ile bir setin başka bir setin süper kümesi ya da doğru süper kümesi olup olmadığını kontrol edebilirsiniz.
