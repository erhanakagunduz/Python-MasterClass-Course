### Python Simetrik Farkı  
Özet: Bu eğitimde, Python'da iki veya daha fazla set arasındaki simetrik farkı nasıl bulacağınızı öğreneceksiniz.

**Setlerin Simetrik Farkına Giriş**  
İki setin simetrik farkı, ya birinci sette ya da ikinci sette bulunan ancak bu setlerin kesişiminde yer almayan öğelerin bulunduğu bir settir.

Varsayalım ki, aşağıdaki `s1` ve `s2` setleriniz var:

```python
s1 = {'Python', 'Java', 'C++'}
s2 = {'C#', 'Java', 'C++'}
```

`s1` ve `s2` setlerinin simetrik farkı aşağıdaki seti döndürür:

```python
{'C#', 'Python'}
```

Çıktıdan açıkça görebileceğiniz gibi, döndürülen setteki öğeler ya `s1` ya da `s2` setinde bulunuyor, ancak kesişimlerinde yer almıyor.

Python'da, iki veya daha fazla set arasındaki simetrik farkı `symmetric_difference()` metodu veya simetrik fark operatörü (^) kullanarak bulabilirsiniz.

### `symmetric_difference()` Metodu ile Setler Arasındaki Simetrik Farkı Bulma
Set türü, iki veya daha fazla set arasındaki simetrik farkı döndüren `symmetric_difference()` metoduna sahiptir:

```python
new_set = set1.symmetric_difference(set2, set3, ...)
```

Örneğin, aşağıdaki kod s1 ve s2 setlerinin simetrik farkını bulur:

```python
s1 = {'Python', 'Java', 'C++'}
s2 = {'C#', 'Java', 'C++'}

s = s1.symmetric_difference(s2)

print(s)
```

Çıktı:

```
{'C#', 'Python'}
```

` symmetric_difference()` metodu, yeni bir set döndürür ve orijinal setleri değiştirmez.

### Simetrik Fark Operatörü (^) ile Setler Arasındaki Simetrik Farkı Bulma
` symmetric_difference()` metodunun dışında, iki veya daha fazla set arasındaki simetrik farkı bulmak için simetrik fark operatörü (^) de kullanılabilir:

```python
new_set = set1 ^ set2 ^ ...
```

Aşağıdaki örnek, simetrik fark operatörünü (^) s1 ve s2 setlerine uygulamayı gösterir:

```python
s1 = {'Python', 'Java', 'C++'}
s2 = {'C#', 'Java', 'C++'}

s = s1 ^ s2

print(s)
```

Çıktı:

```
{'Python', 'C#'}
```

### `symmetric_difference()` Metodu vs. Simetrik Fark Operatörü (^)
` symmetric_difference()` metodu, stringler, listeler veya sözlükler gibi bir veya daha fazla iterable kabul edebilir.

Eğer iterable set değilse, metod, simetrik fark işlemi yapılmadan önce bunları setlere dönüştürür.

Aşağıdaki örnek, bir set ve bir liste arasındaki simetrik farkı bulmak için `symmetric_difference()` metodunu nasıl kullanacağınızı gösterir:

```python
scores = {7, 8, 9}
ratings = [8, 9, 10]
new_set = scores.symmetric_difference(ratings)

print(new_set)
```

Çıktı:

```
{10, 7}
```

Ancak, simetrik fark operatörü (^) yalnızca setler için geçerlidir. Eğer iterable'lar set değilse, bir hata alırsınız. Örneğin:

```python
scores = {7, 8, 9}
ratings = [8, 9, 10]
new_set = scores ^ ratings

print(new_set)
```

Hata:

```
TypeError: unsupported operand type(s) for ^: 'set' and 'list'
```

Simetrik farkı bulmak için `symmetric_difference()` metodunu veya simetrik fark operatörünü (^) kullanabilirsiniz.
