### Python Set Farkına Giriş

İki set arasındaki fark, birinci setten ikinci sette bulunmayan öğeleri içeren yeni bir set oluşturur.

Aşağıdaki gibi `s1` ve `s2` setleriniz olduğunu varsayalım:

```python
s1 = {'Python', 'Java', 'C++'}
s2 = {'C#', 'Java', 'C++'}
```

`s1` ve `s2` setleri arasındaki fark, aşağıdaki gibi bir set oluşturur ve tek bir öğe içerir:

```python
{'Python'}
```

Çünkü yalnızca `'Python'` öğesi, birinci setten (s1) ikinci sette (s2) bulunmamaktadır.

Set farkı komütatif değildir. Yani, `s2` ve `s1` setlerinin farkı şu seti döndürür:

```python
{'C#'}
```

Python'da, setler arasındaki farkı bulmak için `difference()` metodunu veya set farkı operatörü `(-)` kullanabilirsiniz.

1) **Python Set `difference()` metodunu kullanarak setler arasındaki farkı bulmak**  
Set türü, iki veya daha fazla set arasındaki farkı döndüren `difference()` metoduna sahiptir:

```python
set1.difference(s2, s3, ...)
```

Örneğin, `s1` ve `s2` setleri arasındaki farkı bulmak için `difference()` metodunu kullanabilirsiniz:

```python
s1 = {'Python', 'Java', 'C++'}
s2 = {'C#', 'Java', 'C++'}
s = s1.difference(s2)

print(s)
```

Çıktı:

```
{'Python'}
```

Ve bu örnek, `s2` ve `s1` setleri arasındaki farkı bulmak için `difference()` metodunun nasıl kullanılacağını göstermektedir:

```python
s1 = {'Python', 'Java', 'C++'}
s2 = {'C#', 'Java', 'C++'}
s = s2.difference(s1)

print(s)
```

Çıktı:

```
{'C#'}
```

Not: `difference()` metodu yeni bir set döndürür. Orijinal setleri değiştirmez.

2) **Python set farkı operatörü `(-)` ile setler arasındaki farkı bulmak**  
`difference()` metodunun yanı sıra, Python, setler arasındaki farkı bulmak için `(-)` set farkı operatörünü de sağlar.

```python
s = s1 - s2
```

Aşağıdaki örnek, `(-)` fark operatörünü kullanarak `s1` ve `s2` setleri arasındaki farkı bulmaktadır:

```python
s1 = {'Python', 'Java', 'C++'}
s2 = {'C#', 'Java', 'C++'}
s = s1 - s2

print(s)
```

Ve bu örnek, `(-)` set farkı operatörünü kullanarak `s2` ve `s1` arasındaki farkı döndüren bir kullanım gösterir:

```python
s1 = {'Python', 'Java', 'C++'}
s2 = {'C#', 'Java', 'C++'}
s = s2 - s1

print(s)
```

Çıktı:

```
{'C#'}
```

**`difference()` metodu vs. set farkı operatörü `(-)`**  
`difference()` metodu bir veya daha fazla iterable (örneğin, stringler, listeler, sözlükler) kabul edebilir, ancak `(-)` set farkı operatörü yalnızca setleri kabul eder.

`difference()` metoduna iterable (örneğin, liste) geçirdiğinizde, önce iterables'ı setlere dönüştürür ve sonra fark işlemini yapar.

Aşağıdaki örnekte, `difference()` metodunun liste ile nasıl kullanılacağını gösterilmektedir:

```python
scores = {7, 8, 9}
numbers = [9, 10]
new_scores = scores.difference(numbers)

print(new_scores)
```

Ancak, `(-)` fark operatörü ile iterable kullanırsanız hata alırsınız:

```python
scores = {7, 8, 9}
numbers = [9, 10]
new_scores = scores - numbers

print(new_scores)
```

Hata:

```
TypeError: unsupported operand type(s) for -: 'set' and 'list'
```

**Özet**  
İki set arasındaki fark, birinci sette olup ikinci sette bulunmayan öğeleri içeren yeni bir setle sonuçlanır.  
Setler arasındaki farkı bulmak için `difference()` metodunu veya set farkı operatörü `(-)`'yi kullanabilirsiniz.
