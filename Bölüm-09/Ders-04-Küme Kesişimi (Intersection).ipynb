### Python Set Kesişimi

**Set Kesişiminin Kullanım Alanları:**
Set kesişimi, birçok faydalı uygulamaya sahiptir. Örneğin, iki arkadaşın sosyal ağ uygulamasında ortak favorilerini bulmak veya iki veya daha fazla çalışanın İnsan Kaynakları (HR) uygulamasında ortak becerilerini aramak için kullanılabilir.

Python'da, iki veya daha fazla seti kesiştirmek için `intersection()` metodunu veya `&` kesişim operatörünü kullanabilirsiniz.

---

### Python Set intersection() Metodu ile İki veya Daha Fazla Seti Kesiştirme

Aşağıdaki örnek, iki veya daha fazla seti kesiştirmek için `intersection()` metodunun nasıl kullanılacağını gösterir:

```python
new_set = set1.intersection(set2, set3, ...)
```

Aşağıdaki örnek, `s1` ve `s2` setlerini kesiştirmek için `intersection()` metodunun nasıl kullanılacağını gösterir:

```python
s1 = {'Python', 'Java', 'C++'}
s2 = {'C#', 'Java', 'C++'}

s = s1.intersection(s2)

print(s)
```

**Çıktı:**

```
{'C++', 'Java'}
```

---

### Python Set Kesişimi (&) Operatörü ile İki veya Daha Fazla Seti Kesiştirme

Python, iki veya daha fazla seti kesiştirmek için `&` kesişim operatörünü de sağlar:

```python
new_set = s1 & s2 & s3 & ...
```

Aşağıdaki örnek, `s1` ve `s2` setlerini kesiştirmek için `&` operatörünün nasıl kullanılacağını gösterir:

```python
s1 = {'Python', 'Java', 'C++'}
s2 = {'C#', 'Java', 'C++'}

s = s1 & s2

print(s)
```

**Çıktı:**

```
{'C++', 'Java'}
```

---

### `intersection()` Metodu ile `&` Operatörü Arasındaki Farklar

- **`intersection()` metodu**, herhangi bir iterable (string, liste, sözlük gibi) ile çalışabilir, ancak **`&` operatörü sadece setlerle çalışır**.
- Eğer bir iterable (`list`, `string`, vb.) `intersection()` metoduna geçilirse, önce bu iterable bir set'e dönüştürülür ve sonra kesişim yapılır.
- Ancak **`&` operatörü**, iterable'lar ile kullanıldığında hata verir.

Aşağıdaki örnek, bir set ile bir listeyi kesiştirmek için `intersection()` metodunun nasıl kullanılacağını gösterir:

```python
numbers = {1, 2, 3}
scores = [2, 3, 4]

numbers = numbers.intersection(scores)

print(numbers)
```

**Çıktı:**

```
{2, 3}
```

Ancak, aynı işlemi `&` operatörü ile yaparsanız hata alırsınız:

```python
numbers = {1, 2, 3}
scores = [2, 3, 4]

numbers = numbers & scores

print(numbers)
```

**Hata:**

```
TypeError: unsupported operand type(s) for &: 'set' and 'list'
```
