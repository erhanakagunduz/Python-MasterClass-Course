### Python Set Birleşimi

#### Özet:
Bu öğreticide, Python `union()` fonksiyonu veya set birleşim operatörü (`|`) kullanarak iki veya daha fazla setin birleşimini nasıl alacağınızı öğreneceksiniz.

---

### Set Birleşimi'ne Giriş
İki setin birleşimi, her iki setten de ayrık öğeleri içeren yeni bir set döndüren bir işlemdir.

Diyelim ki aşağıdaki setlere sahipsiniz:

```python
s1 = {'Python', 'Java'}
s2 = {'C#', 'Java'}
```

`s1` ve `s2` setlerinin birleşimi şu seti döndürecektir:

```
{'Java', 'Python', 'C#'}
```

Birleşim, her iki setin tüm öğelerini içerir, ancak her öğe yalnızca bir kez yer alır.

### `union()` Metodu ile Set Birleşimi

Python'da iki veya daha fazla seti birleştirmek için `union()` metodunu kullanırsınız:

```python
new_set = set.union(another_set, ...)
```

Aşağıdaki örnek, `s1` ve `s2` setlerinin birleşimini nasıl alacağınızı gösterir:

```python
s1 = {'Python', 'Java'}
s2 = {'C#', 'Java'}

s = s1.union(s2)

print(s)
```

**Çıktı:**

```
{'Python', 'Java', 'C#'}
```

---

### `|` Operatörü ile Set Birleşimi

Python, iki seti birleştirmek için `|` operatörünü de sağlar:

```python
new_set = set1 | set2
```

Bu operatör, `set1` ve `set2`'den ayrık öğeler içeren yeni bir set döndürür.

Aşağıdaki örnek, `|` operatörünü kullanarak `s1` ve `s2` setlerinin birleşimini nasıl alacağınızı gösterir:

```python
s1 = {'Python', 'Java'}
s2 = {'C#', 'Java'}

s = s1 | s2

print(s)
```

**Çıktı:**

```
{'Java', 'C#', 'Python'}
```

---

### `union()` Metodu vs. Set Birleşim Operatörü

- `union()` metodu bir veya daha fazla iterable alır, bu iterablaları setlere dönüştürür ve birleşimi gerçekleştirir.
  
Aşağıdaki örnek, `union()` metoduna bir liste geçirerek nasıl birleşim yapacağınızı gösterir:

```python
rates = {1, 2, 3}
ranks = [2, 3, 4]

ratings = rates.union(ranks)

print(ratings)
```

**Çıktı:**

```
{1, 2, 3, 4}
```

- Ancak, `|` operatörü yalnızca setler için geçerlidir, iterable'lar için değil.

Aşağıdaki örnek bir hata verecektir:

```python
rates = {1, 2, 3}
ranks = [2, 3, 4]

ratings = rates | ranks
```

**Hata:**

```
TypeError: unsupported operand type(s) for |: 'set' and 'list'
```

Sonuç olarak, `union()` metodu iterable'ları kabul ederken, `|` operatörü yalnızca setleri kabul eder.

---

### Özet

- İki veya daha fazla setin birleşimi, her iki setten de ayrık değerler döndürür.
- Bir veya daha fazla seti birleştirmek için `union()` metodunu veya set birleşim operatörünü (`|`) kullanabilirsiniz.
- `union()` metodu bir veya daha fazla iterable alırken, `|` operatörü yalnızca setleri kabul eder.
