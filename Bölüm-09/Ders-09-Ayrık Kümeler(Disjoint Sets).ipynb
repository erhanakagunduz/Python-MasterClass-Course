## Python Ayrık Setler  
Özet: Bu eğitimde, ayrık setleri ve iki setin ayrık olup olmadığını kontrol etmek için Python'un `isdisjoint()` metodunu nasıl kullanacağınızı öğreneceksiniz.

### Python Ayrık Setler  
İki set, ortak elemanları yoksa ayrık setlerdir. Diğer bir deyişle, iki ayrık setin kesişimi boş bir settir.

Örneğin, {1,3,5} ve {2,4,6} setleri ayrık setlerdir çünkü ortak elemanları yoktur.

Aşağıdaki Venn diyagramı, ayrık setleri gösterir:

````python
from IPython.display import Image

# Aynı dizindeki bir resmi görüntüleme
Image(filename="Python-disjoint-sets.png")
````

Python'da, iki setin ayrık olup olmadığını kontrol etmek için `isdisjoint()` metodunu kullanırsınız:

```python
set_a.isdisjoint(set_b)
```

`isdisjoint()` metodu, `set_a` ve `set_b` ayrık ise `True` döner. Aksi takdirde `False` döner.

`isdisjoint()` metodu, sadece setleri değil, herhangi bir yinelemeli (iterable) veri tipini de kabul eder. 

Bir liste, demet (tuple) veya sözlük (dictionary) geçirirseniz, `isdisjoint()` metodu önce bunu bir sete dönüştürür ve sonra kontrol eder.

### Python `isdisjoint()` Metodu Örnekleri  
Aşağıdaki örnek, `isdisjoint()` metodunu kullanarak `odd_numbers` ve `even_numbers` setlerinin ayrık olup olmadığını kontrol eder:

```python
odd_numbers = {1, 3, 5}
even_numbers = {2, 4, 6}

result = odd_numbers.isdisjoint(even_numbers)

print(result)
```

Çıktı:

```
True
```

`odd_numbers` setindeki hiçbir eleman `even_numbers` setinde bulunmadığı için `isdisjoint()` metodu `True` döner.

Aşağıdaki örnek, `isdisjoint()` metodunu kullanarak `letters` ve `alphanumerics` setlerinin ayrık olup olmadığını kontrol eder:

```python
letters = {'A', 'B', 'C'}
alphanumerics = {'A', 1, 2}

result = letters.isdisjoint(alphanumerics)

print(result)
```

Çıktı:

```
False
```

`alphanumerics` setindeki 'A' harfi `letters` setinde de bulunduğundan, `isdisjoint()` metodu `False` döner.

Aşağıdaki örnek, bir set yerine bir listeyi `isdisjoint()` metoduna geçirir:

```python
letters = {'A', 'B', 'C'}
result = letters.isdisjoint([1, 2, 3])

print(result)
```

Çıktı:

```
True
```

### Özet  
İki set ayrık ise, ortak elemanları yok demektir.  
İki setin ayrık olup olmadığını kontrol etmek için Python'daki `isdisjoint()` metodunu kullanın.
